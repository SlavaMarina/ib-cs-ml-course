```python
# Basic flashcard infrastructure
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML

# Global state
class FlashcardSession:
    def __init__(self):
        self.cards_seen = 0
        self.cards_hard = []

session = FlashcardSession()

def make_flashcard_set(cards, set_name='Set'):
    # Creates an interactive set of flashcards
    state = {'idx': 0, 'show_answer': False}

    front_area = widgets.HTML()
    back_area = widgets.HTML()

    show_btn = widgets.Button(description='👁 Show answer', button_style='primary')
    next_btn = widgets.Button(description='➡ Next card', button_style='success')
    prev_btn = widgets.Button(description='⬅ Previous', button_style='')
    hard_btn = widgets.Button(description='⚠ Hard', button_style='warning')

    counter = widgets.HTML()

    def update():
        idx = state['idx']
        card = cards[idx]
        front_area.value = (
            f"<div style='background:#FFF8DC;padding:20px;border-radius:10px;"
            f"border:3px solid #DAA520;margin:10px 0;'>"
            f"<h3 style='color:#8B4513;margin-top:0;'>🎴 Card {idx+1}/{len(cards)} · {set_name}</h3>"
            f"<div style='font-size:18px;'><b>Q:</b> {card['front']}</div>"
            f"</div>"
        )
        if state['show_answer']:
            back_area.value = (
                f"<div style='background:#E8F5E9;padding:20px;border-radius:10px;"
                f"border:3px solid #4CAF50;margin:10px 0;'>"
                f"<div style='font-size:16px;'><b>A:</b> {card['back']}</div>"
                f"</div>"
            )
        else:
            back_area.value = ""
        counter.value = f"<i>Sequence: {idx+1}/{len(cards)}</i>"

    def on_show(b):
        state['show_answer'] = not state['show_answer']
        update()

    def on_next(b):
        state['idx'] = (state['idx'] + 1) % len(cards)
        state['show_answer'] = False
        session.cards_seen += 1
        update()

    def on_prev(b):
        state['idx'] = (state['idx'] - 1) % len(cards)
        state['show_answer'] = False
        update()

    def on_hard(b):
        card_id = f"{set_name}-{state['idx']+1}"
        if card_id not in session.cards_hard:
            session.cards_hard.append(card_id)
            print(f"⚠ Marked as hard: {card_id}")

    show_btn.on_click(on_show)
    next_btn.on_click(on_next)
    prev_btn.on_click(on_prev)
    hard_btn.on_click(on_hard)

    controls = widgets.HBox([prev_btn, show_btn, next_btn, hard_btn])
    update()

    display(front_area, back_area, controls, counter)

print("✅ Flashcard infrastructure ready!")
```
