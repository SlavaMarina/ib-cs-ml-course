### ⚠️ How it works

1. Run the first cell with imports
2. Each question has **4 answer options** (radio buttons)
3. Click **"Check"** — you will see the correct answer + explanation
4. At the end, you get a total score (% correct)

```python
# Setup — run this cell first
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output

# Global score tracker
global_score = {'correct': 0, 'attempted': 0, 'by_topic': {}}

def make_question(qid, topic, question_text, options, correct_idx, explanation):
    # Creates an interactive MC question
    radio = widgets.RadioButtons(
        options=[(opt, i) for i, opt in enumerate(options)],
        description='',
        disabled=False,
        layout=widgets.Layout(width='95%')
    )
    check_btn = widgets.Button(description='Check answer', button_style='primary',
                                layout=widgets.Layout(width='200px'))
    output = widgets.Output()

    def on_check(b):
        with output:
            clear_output()
            selected = radio.value
            if selected is None:
                print("⚠️ Please select an option first.")
                return

            global_score['attempted'] += 1
            global_score['by_topic'].setdefault(topic, {'correct': 0, 'total': 0})
            global_score['by_topic'][topic]['total'] += 1

            if selected == correct_idx:
                global_score['correct'] += 1
                global_score['by_topic'][topic]['correct'] += 1
                display(HTML(f'<p style="color: green; font-weight: bold;">✅ CORRECT!</p>'))
            else:
                display(HTML(f'<p style="color: red; font-weight: bold;">❌ INCORRECT.</p>'
                             f'<p>Correct answer: <b>{options[correct_idx]}</b></p>'))
            display(HTML(f'<p style="font-style: italic;">💡 {explanation}</p>'))
            check_btn.disabled = True

    check_btn.on_click(on_check)

    title = widgets.HTML(value=f'<h4>Q{qid} ({topic})</h4><p>{question_text}</p>')
    display(widgets.VBox([title, radio, check_btn, output]))

print("✅ Setup complete. Run the following cells for the questions.")
```
