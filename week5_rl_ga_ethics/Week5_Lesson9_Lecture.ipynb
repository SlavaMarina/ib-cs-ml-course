## 🎮 Часть 1 · A4.3.6 Reinforcement Learning

> **Definition (выучить!):** *Reinforcement learning is a paradigm where an agent learns to make decisions by **interacting with its environment**, taking actions and receiving rewards or penalties, with the goal of maximizing cumulative reward over time.*

### 🧠 Аналогия: ребёнок учится ходить

> Из Baumgarten: *"It can be likened to a toddler learning to walk. Every time the toddler falls over, they learn a little more about how to correctly balance themself next time."*

- **Agent** = ребёнок
- **Environment** = пол + гравитация
- **State** = текущая поза тела
- **Action** = шаг ногой, размах руки
- **Reward** = устойчиво стоит (+) / упал (−)
- **Policy** = "как двигаться при таком положении тела"

---

### 📐 6 ключевых терминов (КРИТИЧНО для IB!)

| Термин | Определение |
|---|---|
| **Agent** | сущность, которая **принимает решения** (модель ML) |
| **Environment** | мир, в котором действует агент |
| **State** | снимок мира в момент времени (что агент **наблюдает**) |
| **Action** | операция, которую агент может выполнить |
| **Reward** | обратная связь (+/-) от среды на действие |
| **Policy** | стратегия "if state X, then action Y" |

## ⚖️ Часть 2 · Q-learning + Exploration vs Exploitation

### Q-learning — как агент "запоминает"

> Самый популярный RL-алгоритм. Идея: создать **таблицу** `Q[state][action]` = ожидаемая награда от выбора `action` в `state`.

**Пример Q-таблицы (Baumgarten p. 53):**

|  | Action 1 | Action 2 | Action 3 | Action 4 |
|---|---|---|---|---|
| **State 1** | -50 | 0 | **10** ✅ | 0 |
| **State 2** | 10 | **20** ✅ | 0 | 10 |
| **State 3** | 0 | -10 | 0 | **50** ✅ |

> Из таблицы агент видит: "в State 1 лучше всего делать Action 3 (reward +10), а Action 1 (-50) — избегать".

> 💡 **Когда Q-таблица слишком большая** (миллионы states × actions) → используем **нейросеть** для аппроксимации → это **Deep Q-Network (DQN)**.

---

### 🎯 Exploration vs Exploitation — **trade-off**

> **Exploration** = пробовать **новые** действия (вдруг найдём что-то лучше?)
> **Exploitation** = использовать **известные хорошие** действия

| Стратегия | Когда применять |
|---|---|
| **Чистый exploration** | в начале обучения — агент ничего не знает |
| **Чистый exploitation** | в конце — агент всё выучил |
| **Epsilon-greedy** | компромисс: с вероятностью **ε** explore, иначе exploit |
| **UCB** (Upper Confidence Bound) | статистическая балансировка |
| **Thompson sampling** | вероятностный подход |

> 💎 **СЕКРЕТ #3:** *"Discuss the exploration-exploitation trade-off"* **[4]** — образец:
> 1. **Pure exploration**: агент учится быстро, но может не использовать найденные хорошие стратегии (1)
> 2. **Pure exploitation**: агент следует первой найденной хорошей стратегии, но может **застрять в локальном максимуме** (1)
> 3. **Epsilon-greedy** balances both: high ε at start, decay over time (1)
> 4. **Optimal balance** зависит от задачи: для безопасности (medical) — больше exploitation; для исследования (game AI) — больше exploration (1)


```python
# Демонстрация: как ε (epsilon) меняется со временем
fig, ax = plt.subplots(figsize=(10, 5))
episodes = np.arange(1, 1001)

# Decay strategies
epsilon_constant = np.ones_like(episodes) * 0.3
epsilon_linear = np.maximum(0.05, 1 - episodes * 0.001)
epsilon_exp = np.maximum(0.05, np.exp(-episodes * 0.003))

ax.plot(episodes, epsilon_constant, label='Constant ε=0.3', linewidth=2)
ax.plot(episodes, epsilon_linear, label='Linear decay', linewidth=2)
ax.plot(episodes, epsilon_exp, label='Exponential decay', linewidth=2)

ax.fill_between(episodes[:200], 0, 1, alpha=0.15, color='orange', label='Exploration phase')
ax.fill_between(episodes[700:], 0, 1, alpha=0.15, color='green', label='Exploitation phase')

ax.set_xlabel('Episode (training step)', fontsize=11)
ax.set_ylabel('ε (probability of random action)', fontsize=11)
ax.set_title('Epsilon-greedy: agent учится → ε уменьшается → больше exploitation',
             fontsize=12, fontweight='bold')
ax.legend(); ax.grid(alpha=0.3)
plt.tight_layout(); plt.show()
```

## 🧬 Часть 3 · A4.3.7 Genetic Algorithms

> **Definition (выучить!):** *A genetic algorithm imitates the concept of **survival of the fittest** and evolution by testing a population of possible solutions, using properties from the **best-performing solutions** to create a new population, and repeating until a suitable solution emerges.*

> 🚨 **Важно (Baumgarten):** GA — это **НЕ** supervised / unsupervised / reinforcement.
> Это **evolutionary algorithm**. На экзамене не путайте.

### 🧬 6 шагов алгоритма

| Шаг | Что делает |
|---|---|
| 1. **Initialization** | создаём случайную **populations** (например, 500 случайных решений) |
| 2. **Fitness Evaluation** | каждое решение оцениваем функцией fitness |
| 3. **Selection** | отбираем "родителей" — лучшие решения с большей вероятностью |
| 4. **Crossover** | "скрещиваем" родителей → создаём детей |
| 5. **Mutation** | случайно изменяем небольшую часть детей |
| 6. **Replacement** | новое поколение заменяет старое |
| **(Repeat)** | повторяем 2-6 до **termination criteria** |

> 💎 **СЕКРЕТ #4:** часто на IB: *"Outline the role of crossover in genetic algorithms"* **[2]**.
> Образец: *"Crossover combines genetic information from two parent solutions to create a new offspring solution that inherits properties from both. This allows good traits from successful solutions to combine, potentially producing even better solutions."*


```python
# Простейшая реализация GA на примере TSP (Traveling Salesperson)
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# 10 городов в случайных координатах
N_CITIES = 10
cities = np.random.rand(N_CITIES, 2) * 100

def route_distance(route):
    # Расстояние замкнутого маршрута
    dist = 0
    for i in range(len(route)):
        a = cities[route[i]]
        b = cities[route[(i+1) % len(route)]]
        dist += np.linalg.norm(a - b)
    return dist

def fitness(route):
    # Fitness — обратное расстоянию (короче маршрут → выше fitness)
    return 1 / route_distance(route)

# Сгенерируем 5 случайных маршрутов как начальную популяцию
population_size = 100
population = [np.random.permutation(N_CITIES) for _ in range(population_size)]

# Вычислим fitness первого поколения
initial_best = min(population, key=route_distance)
print(f"🎲 Случайный лучший маршрут (поколение 0): дистанция = {route_distance(initial_best):.1f}")

# Визуализация
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
for ax, route, title in [
    (axes[0], initial_best, 'Случайный маршрут (gen 0)'),
    (axes[1], initial_best, 'Случайный маршрут (gen 0)')  # потом обновим
]:
    closed = list(route) + [route[0]]
    pts = cities[closed]
    ax.plot(pts[:, 0], pts[:, 1], 'o-', markersize=10)
    ax.set_title(f'{title}\nDistance = {route_distance(route):.1f}', fontweight='bold')
    for i, (x, y) in enumerate(cities):
        ax.annotate(str(i), (x, y), fontsize=10, xytext=(3, 3), textcoords='offset points')
    ax.set_aspect('equal'); ax.grid(alpha=0.3)
plt.tight_layout(); plt.show()
```

```python
# Запустим GA на 100 поколений
def selection(population, k=3):
    # Tournament selection: выбираем k случайных, берём лучшего
    competitors = np.random.choice(len(population), k, replace=False)
    return min([population[i] for i in competitors], key=route_distance)

def crossover(parent1, parent2):
    # One-point crossover для permutation (адаптированный)
    n = len(parent1)
    point = np.random.randint(1, n - 1)
    child = list(parent1[:point])
    for city in parent2:
        if city not in child:
            child.append(city)
    return np.array(child)

def mutate(route, rate=0.05):
    route = route.copy()
    if np.random.random() < rate:
        # Меняем местами 2 случайных города
        i, j = np.random.choice(len(route), 2, replace=False)
        route[i], route[j] = route[j], route[i]
    return route

# Цикл эволюции
best_distances = []
generations = 100

for gen in range(generations):
    new_population = []
    for _ in range(population_size):
        p1 = selection(population)
        p2 = selection(population)
        child = crossover(p1, p2)
        child = mutate(child)
        new_population.append(child)
    population = new_population
    best = min(population, key=route_distance)
    best_distances.append(route_distance(best))

best_route = min(population, key=route_distance)
print(f"✅ После {generations} поколений: лучшая дистанция = {route_distance(best_route):.1f}")
print(f"   Улучшение: {route_distance(initial_best):.1f} → {route_distance(best_route):.1f}")
print(f"   ({(1 - route_distance(best_route)/route_distance(initial_best))*100:.1f}% сокращение)")
```

```python
# Визуализация эволюции
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Лучший начальный маршрут
closed = list(initial_best) + [initial_best[0]]
pts = cities[closed]
axes[0].plot(pts[:, 0], pts[:, 1], 'o-', markersize=10, color='red')
axes[0].set_title(f'Generation 0: {route_distance(initial_best):.1f}',
                  fontweight='bold', color='darkred')
for i, (x, y) in enumerate(cities):
    axes[0].annotate(str(i), (x, y), fontsize=10, xytext=(3, 3), textcoords='offset points')

# Лучший финальный маршрут
closed = list(best_route) + [best_route[0]]
pts = cities[closed]
axes[1].plot(pts[:, 0], pts[:, 1], 'o-', markersize=10, color='green')
axes[1].set_title(f'Generation {generations}: {route_distance(best_route):.1f}',
                  fontweight='bold', color='darkgreen')
for i, (x, y) in enumerate(cities):
    axes[1].annotate(str(i), (x, y), fontsize=10, xytext=(3, 3), textcoords='offset points')

# Кривая улучшения
axes[2].plot(best_distances, linewidth=2, color='steelblue')
axes[2].set_xlabel('Поколение'); axes[2].set_ylabel('Лучшая дистанция')
axes[2].set_title('Эволюция fitness по поколениям', fontweight='bold')
axes[2].grid(alpha=0.3)

for ax in axes[:2]:
    ax.set_aspect('equal'); ax.grid(alpha=0.3)

plt.tight_layout(); plt.show()
```

|---|---|---|

### 💡 GA vs другие подходы — когда использовать?

| Используйте GA если... | НЕ используйте GA если... |
|---|---|
| Огромное пространство решений (TSP, scheduling) | Простые задачи (можно решить аналитически) |
| Нет четкой математической модели | Точное решение **критично** |
| "Достаточно хорошее" решение приемлемо | Требуется глобальный оптимум |
| Множество локальных максимумов | Малое пространство решений |

> 💎 **СЕКРЕТ #5:** **Common mistake**: путать GA с RL. Различие:
> - **RL**: один агент учится **со временем**
> - **GA**: **популяция** решений эволюционирует через "поколения"
