## 🎮 Part 1 · A4.3.6 Reinforcement Learning

> **Definition (memorise this):** *Reinforcement learning is a paradigm where an agent learns to make decisions by **interacting with its environment**, taking actions and receiving rewards or penalties, with the goal of maximizing cumulative reward over time.*

### 🧠 Analogy: a toddler learning to walk

> From Baumgarten: *"It can be likened to a toddler learning to walk. Every time the toddler falls over, they learn a little more about how to correctly balance themself next time."*

- **Agent** = child
- **Environment** = floor + gravity
- **State** = current body position
- **Action** = moving a foot, swinging an arm
- **Reward** = standing steadily (+) / falling (−)
- **Policy** = "how to move in this body position"

---

### 📐 6 key terms (CRITICAL for IB!)

| Term | Definition |
|---|---|
| **Agent** | the entity that **makes decisions** (the ML model) |
| **Environment** | the world in which the agent acts |
| **State** | a snapshot of the world at a point in time (what the agent **observes**) |
| **Action** | an operation the agent can perform |
| **Reward** | positive or negative feedback from the environment after an action |
| **Policy** | the strategy: "if state X, then action Y" |

> 💎 **SECRET #1:** *"Identify and describe the 5 components of reinforcement learning"* **[5]** is a standard question. Learn these **by heart**: Agent · Environment · State · Action · Reward. Policy is an extra strong term.


```python
# Visualisation: agent-environment loop
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

fig, ax = plt.subplots(figsize=(11, 7))

# Agent box
agent_box = mpatches.FancyBboxPatch((0.1, 0.55), 0.25, 0.20,
                                      boxstyle="round,pad=0.02",
                                      facecolor='#4ECDC4', edgecolor='black', linewidth=2)
ax.add_patch(agent_box)
ax.text(0.225, 0.65, 'AGENT\n(ML model)', ha='center', va='center',
        fontsize=14, fontweight='bold')

# Environment box
env_box = mpatches.FancyBboxPatch((0.65, 0.55), 0.25, 0.20,
                                    boxstyle="round,pad=0.02",
                                    facecolor='#FFD93D', edgecolor='black', linewidth=2)
ax.add_patch(env_box)
ax.text(0.775, 0.65, 'ENVIRONMENT\n(world)', ha='center', va='center',
        fontsize=14, fontweight='bold')

# Arrow: action from agent to environment
ax.annotate('', xy=(0.65, 0.70), xytext=(0.35, 0.70),
            arrowprops=dict(arrowstyle='->', lw=3, color='steelblue'))
ax.text(0.50, 0.74, 'ACTION', ha='center', fontsize=12, fontweight='bold',
        color='steelblue')

# Arrows: state + reward from environment to agent
ax.annotate('', xy=(0.35, 0.62), xytext=(0.65, 0.62),
            arrowprops=dict(arrowstyle='->', lw=2.5, color='darkred'))
ax.text(0.50, 0.58, 'REWARD', ha='center', fontsize=11, color='darkred')

ax.annotate('', xy=(0.35, 0.58), xytext=(0.65, 0.58),
            arrowprops=dict(arrowstyle='->', lw=2.5, color='darkgreen'))
ax.text(0.50, 0.54, 'STATE', ha='center', fontsize=11, color='darkgreen')

# Policy
ax.text(0.225, 0.40, '📋 POLICY\n"if state X → action Y"',
        ha='center', fontsize=11, style='italic',
        bbox=dict(boxstyle="round,pad=0.5", facecolor='lightyellow', edgecolor='gray'))

# Learning loop
ax.text(0.50, 0.20,
        '🔄 LEARNING LOOP\n1. Agent observes state\n2. Agent picks action (using policy)\n'
        '3. Environment updates → new state + reward\n4. Agent updates policy\n5. Repeat...',
        ha='center', fontsize=10,
        bbox=dict(boxstyle="round,pad=0.5", facecolor='lightblue', edgecolor='gray'))

ax.set_xlim(0, 1); ax.set_ylim(0, 1)
ax.axis('off')
ax.set_title('Reinforcement Learning — core interaction loop',
             fontsize=14, fontweight='bold')
plt.tight_layout(); plt.show()
```

## ⚖️ Part 2 · Q-learning + Exploration vs Exploitation

### Q-learning — how the agent "remembers"

> This is one of the best-known RL algorithms. The idea is to create a **table**: `Q[state][action]` = expected reward for choosing `action` in `state`.

**Example Q-table (Baumgarten p. 53):**

|  | Action 1 | Action 2 | Action 3 | Action 4 |
|---|---|---|---|---|
| **State 1** | -50 | 0 | **10** ✅ | 0 |
| **State 2** | 10 | **20** ✅ | 0 | 10 |
| **State 3** | 0 | -10 | 0 | **50** ✅ |

> From the table, the agent sees: "in State 1, Action 3 is best (reward +10), while Action 1 (-50) should be avoided".

> 💡 **When the Q-table becomes too large** (millions of states × actions), we use a **neural network** to approximate it. This is a **Deep Q-Network (DQN)**.

---

### 🎯 Exploration vs Exploitation — **trade-off**

> **Exploration** = try **new** actions (maybe there is something better?)
> **Exploitation** = use **known good** actions

| Strategy | When to use it |
|---|---|
| **Pure exploration** | at the start of training, when the agent knows nothing |
| **Pure exploitation** | near the end, when the agent has already learnt enough |
| **Epsilon-greedy** | compromise: with probability **ε**, explore; otherwise exploit |
| **UCB** (Upper Confidence Bound) | statistical balancing |
| **Thompson sampling** | probabilistic approach |

> 💎 **SECRET #3:** *"Discuss the exploration-exploitation trade-off"* **[4]** — sample structure:
> 1. **Pure exploration**: the agent learns quickly, but may fail to use good strategies it has already found (1)
> 2. **Pure exploitation**: the agent follows the first good strategy it finds, but may **get stuck in a local maximum** (1)
> 3. **Epsilon-greedy** balances both: high ε at the start, then decay over time (1)
> 4. **The optimal balance** depends on the task: for safety-critical tasks (medical), use more exploitation; for discovery tasks (game AI), use more exploration (1)


```python
# Demo: how ε (epsilon) changes over time
fig, ax = plt.subplots(figsize=(10, 5))
episodes = np.arange(1, 1001)

# Decay strategies
epsilon_constant = np.ones_like(episodes) * 0.3
epsilon_linear = np.maximum(0.05, 1 - episodes * 0.001)
epsilon_exp = np.maximum(0.05, np.exp(-episodes * 0.003))

ax.plot(episodes, epsilon_constant, label='Constant ε=0.3', linewidth=2)
ax.plot(episodes, epsilon_linear, label='Linear decay', linewidth=2)
ax.plot(episodes, epsilon_exp, label='Exponential decay', linewidth=2)

ax.fill_between(episodes[:200], 0, 1, alpha=0.15, color='orange', label='Exploration phase')
ax.fill_between(episodes[700:], 0, 1, alpha=0.15, color='green', label='Exploitation phase')

ax.set_xlabel('Episode (training step)', fontsize=11)
ax.set_ylabel('ε (probability of random action)', fontsize=11)
ax.set_title('Epsilon-greedy: agent learns → ε decreases → more exploitation',
             fontsize=12, fontweight='bold')
ax.legend(); ax.grid(alpha=0.3)
plt.tight_layout(); plt.show()
```

## 🧬 Part 3 · A4.3.7 Genetic Algorithms

> **Definition (memorise this):** *A genetic algorithm imitates the concept of **survival of the fittest** and evolution by testing a population of possible solutions, using properties from the **best-performing solutions** to create a new population, and repeating until a suitable solution emerges.*

> 🚨 **Important (Baumgarten):** GA is **NOT** supervised / unsupervised / reinforcement learning.
> It is an **evolutionary algorithm**. Do not confuse these in the exam.

### 🧬 6 algorithm steps

| Step | What it does |
|---|---|
| 1. **Initialization** | create a random **population** (for example, 500 random solutions) |
| 2. **Fitness Evaluation** | evaluate each solution using a fitness function |
| 3. **Selection** | select "parents"; better solutions are more likely to be chosen |
| 4. **Crossover** | combine parents to create children |
| 5. **Mutation** | randomly change a small part of the children |
| 6. **Replacement** | the new generation replaces the old one |
| **(Repeat)** | repeat steps 2-6 until the **termination criteria** are met |

> 💎 **SECRET #4:** IB often asks: *"Outline the role of crossover in genetic algorithms"* **[2]**.
> Sample answer: *"Crossover combines genetic information from two parent solutions to create a new offspring solution that inherits properties from both. This allows good traits from successful solutions to combine, potentially producing even better solutions."*


```python
# Simple GA implementation for TSP (Traveling Salesperson Problem)
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# 10 cities at random coordinates
N_CITIES = 10
cities = np.random.rand(N_CITIES, 2) * 100

def route_distance(route):
    # Distance of a closed route
    dist = 0
    for i in range(len(route)):
        a = cities[route[i]]
        b = cities[route[(i+1) % len(route)]]
        dist += np.linalg.norm(a - b)
    return dist

def fitness(route):
    # Fitness is the inverse of distance (shorter route -> higher fitness)
    return 1 / route_distance(route)

# Generate random routes as the initial population
population_size = 100
population = [np.random.permutation(N_CITIES) for _ in range(population_size)]

# Calculate fitness for the first generation
initial_best = min(population, key=route_distance)
print(f"🎲 Random best route (generation 0): distance = {route_distance(initial_best):.1f}")

# Visualisation
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
for ax, route, title in [
    (axes[0], initial_best, 'Random route (gen 0)'),
    (axes[1], initial_best, 'Random route (gen 0)')  # will update later
]:
    closed = list(route) + [route[0]]
    pts = cities[closed]
    ax.plot(pts[:, 0], pts[:, 1], 'o-', markersize=10)
    ax.set_title(f'{title}\nDistance = {route_distance(route):.1f}', fontweight='bold')
    for i, (x, y) in enumerate(cities):
        ax.annotate(str(i), (x, y), fontsize=10, xytext=(3, 3), textcoords='offset points')
    ax.set_aspect('equal'); ax.grid(alpha=0.3)
plt.tight_layout(); plt.show()
```

```python
# Run GA for 100 generations
def selection(population, k=3):
    # Tournament selection: choose k random routes, then keep the best one
    competitors = np.random.choice(len(population), k, replace=False)
    return min([population[i] for i in competitors], key=route_distance)

def crossover(parent1, parent2):
    # One-point crossover for a permutation (adapted)
    n = len(parent1)
    point = np.random.randint(1, n - 1)
    child = list(parent1[:point])
    for city in parent2:
        if city not in child:
            child.append(city)
    return np.array(child)

def mutate(route, rate=0.05):
    route = route.copy()
    if np.random.random() < rate:
        # Swap two random cities
        i, j = np.random.choice(len(route), 2, replace=False)
        route[i], route[j] = route[j], route[i]
    return route

# Evolution loop
best_distances = []
generations = 100

for gen in range(generations):
    new_population = []
    for _ in range(population_size):
        p1 = selection(population)
        p2 = selection(population)
        child = crossover(p1, p2)
        child = mutate(child)
        new_population.append(child)
    population = new_population
    best = min(population, key=route_distance)
    best_distances.append(route_distance(best))

best_route = min(population, key=route_distance)
print(f"✅ After {generations} generations: best distance = {route_distance(best_route):.1f}")
print(f"   Improvement: {route_distance(initial_best):.1f} → {route_distance(best_route):.1f}")
print(f"   ({(1 - route_distance(best_route)/route_distance(initial_best))*100:.1f}% reduction)")
```

```python
# Visualise evolution
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Best initial route
closed = list(initial_best) + [initial_best[0]]
pts = cities[closed]
axes[0].plot(pts[:, 0], pts[:, 1], 'o-', markersize=10, color='red')
axes[0].set_title(f'Generation 0: {route_distance(initial_best):.1f}',
                  fontweight='bold', color='darkred')
for i, (x, y) in enumerate(cities):
    axes[0].annotate(str(i), (x, y), fontsize=10, xytext=(3, 3), textcoords='offset points')

# Best final route
closed = list(best_route) + [best_route[0]]
pts = cities[closed]
axes[1].plot(pts[:, 0], pts[:, 1], 'o-', markersize=10, color='green')
axes[1].set_title(f'Generation {generations}: {route_distance(best_route):.1f}',
                  fontweight='bold', color='darkgreen')
for i, (x, y) in enumerate(cities):
    axes[1].annotate(str(i), (x, y), fontsize=10, xytext=(3, 3), textcoords='offset points')

# Improvement curve
axes[2].plot(best_distances, linewidth=2, color='steelblue')
axes[2].set_xlabel('Generation'); axes[2].set_ylabel('Best distance')
axes[2].set_title('Fitness evolution by generation', fontweight='bold')
axes[2].grid(alpha=0.3)

for ax in axes[:2]:
    ax.set_aspect('equal'); ax.grid(alpha=0.3)

plt.tight_layout(); plt.show()
```

### 🎯 Real-world uses of GA (for IB Section B)

| Scenario | What we optimise | Fitness function |
|---|---|---|
| **TSP** (courier route) | route between N cities | inverse of total distance |
| **School timetable** | minimum conflicts | number of satisfied constraints |
| **Wind farm layout** | position of wind turbines | energy production − cost |
| **Aircraft design** | wing shape | lift / drag |
| **Portfolio optimisation** | asset weights in a portfolio | Sharpe ratio (return/risk) |

### 💡 GA vs other approaches — when should we use it?

| Use GA if... | Do NOT use GA if... |
|---|---|
| The solution space is huge (TSP, scheduling) | The task is simple and can be solved analytically |
| There is no clear mathematical model | An exact solution is **critical** |
| A "good enough" solution is acceptable | A guaranteed global optimum is required |
| There are many local maxima | The solution space is small |

> 💎 **SECRET #5:** **Common mistake**: confusing GA with RL. Difference:
> - **RL**: one agent learns **over time**
> - **GA**: a **population** of solutions evolves through "generations"

### 🎯 8 types of ethical issue (CRITICAL to learn!)

| # | Issue | Core idea | Known case |
|---|---|---|---|
| 1 | **Accountability** | who is responsible for AI errors? | Uber 2018 fatal crash; who was responsible: safety driver, company, developer, or regulator? |
| 2 | **Algorithmic fairness** | algorithms can discriminate | Amazon (2018) stopped an AI recruiting tool because it discriminated against women |
| 3 | **Bias in training data** | bias can be inherited from data | COMPAS recidivism: higher false positives for Black defendants |
| 4 | **Privacy & consent** | data collected or shared without proper consent | Netflix 2006: an anonymised ratings dataset was de-anonymised |
| 5 | **Environmental impact** | training large models can use substantial energy | large model training requires major data-centre energy use |
| 6 | **Security & misuse** | models can be vulnerable or misused | adversarial attacks, prompt injections |
| 7 | **Societal disruption** | jobs and social behaviour can change at scale | automation can displace large numbers of workers |
| 8 | **Transparency / explainability** | decisions may be a "black box" | banking algorithms may deny credit without a clear explanation |

### 📚 Ready case studies for the exam (LEARN THESE!)

> Use these **specific** examples in answers. Examiners value concrete evidence.

#### Case 1: COMPAS (criminal justice)
> ProPublica 2016: the COMPAS algorithm in US courts predicted recidivism risk.
> Finding: **Black** defendants received higher false positive scores than white defendants.
> **Syllabus link:** algorithmic fairness + bias in training data.

#### Case 2: Amazon AI recruiting (2018)
> Amazon stopped an AI recruiting tool.
> Reason: the model was trained on 10 years of CVs, where men dominated the data → **discrimination against women**.
> **Link:** training data bias + algorithmic fairness.

#### Case 3: Netflix de-anonymisation (2006)
> Netflix published an "anonymised" ratings dataset.
> Researchers from UT Austin showed that users could be **identified** by matching the data with IMDb.
> **Link:** privacy violation + insufficient anonymisation.

#### Case 4: Uber self-driving crash (2018)
> An Uber self-driving vehicle killed a pedestrian in Arizona.
> The safety driver received probation; Uber did not receive the same legal punishment.
> **Link:** accountability — who is responsible?

#### Case 5: Deepfakes
> Videos can fake statements by politicians or create non-consensual sexual content.
> **Link:** misinformation + consent + reputation damage.

#### Case 6: Facebook/YouTube echo chambers
> Recommendation algorithms may show users only content they already like → social polarisation.
> **Link:** filter bubbles + manipulation + societal harm.

> 3. **Real example** (1 mark)
