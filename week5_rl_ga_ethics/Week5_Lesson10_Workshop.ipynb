## 🧬 Part 3 · GA in practice — N-Queens problem

> **N-Queens**: place N queens on an N×N board so that they **do not attack** each other.
> This is a **classic** optimisation problem and a strong example for GA.


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Circle

N = 8  # board size
POPULATION_SIZE = 100
GENERATIONS = 200
MUTATION_RATE = 0.1

def random_solution():
    # solution[i] = row where the queen is placed in column i
    return np.random.randint(0, N, size=N)

def fitness(solution):
    # Count the number of NON-attacking queen pairs (max = N*(N-1)/2)
    non_attacking = 0
    for i in range(N):
        for j in range(i+1, N):
            if solution[i] != solution[j] and abs(solution[i] - solution[j]) != abs(i - j):
                non_attacking += 1
    return non_attacking

def selection(population, fitnesses):
    # Tournament selection
    idxs = np.random.choice(len(population), 3, replace=False)
    return population[idxs[np.argmax([fitnesses[i] for i in idxs])]]

def crossover(p1, p2):
    point = np.random.randint(1, N-1)
    return np.concatenate([p1[:point], p2[point:]])

def mutate(sol):
    sol = sol.copy()
    if np.random.random() < MUTATION_RATE:
        idx = np.random.randint(N)
        sol[idx] = np.random.randint(N)
    return sol

np.random.seed(42)
population = [random_solution() for _ in range(POPULATION_SIZE)]
best_fitness_history = []
target = N * (N - 1) // 2  # maximum possible number of non-attacking pairs

print(f"🎯 N = {N}, target fitness = {target}\n")

for gen in range(GENERATIONS):
    fitnesses = [fitness(s) for s in population]
    best_fitness_history.append(max(fitnesses))

    if max(fitnesses) == target:
        print(f"✅ Solution found in generation {gen}!")
        break

    # New generation
    new_pop = []
    for _ in range(POPULATION_SIZE):
        p1 = selection(population, fitnesses)
        p2 = selection(population, fitnesses)
        child = crossover(p1, p2)
        child = mutate(child)
        new_pop.append(child)
    population = new_pop

best_solution = population[np.argmax([fitness(s) for s in population])]
print(f"Best solution: {best_solution}")
print(f"Fitness: {fitness(best_solution)}/{target}")
```

```python
# Visualisation: final board + learning curve
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Board
ax = axes[0]
for i in range(N):
    for j in range(N):
        color = 'lightgray' if (i+j) % 2 == 0 else 'white'
        ax.add_patch(Rectangle((j, i), 1, 1, facecolor=color, edgecolor='black'))

# Queens
for col, row in enumerate(best_solution):
    ax.add_patch(Circle((col + 0.5, row + 0.5), 0.3, color='darkred', zorder=5))
    ax.text(col + 0.5, row + 0.5, '♛', ha='center', va='center',
            fontsize=20, color='gold', zorder=6)

ax.set_xlim(0, N); ax.set_ylim(0, N)
ax.set_aspect('equal')
ax.set_xticks([]); ax.set_yticks([])
ax.set_title(f'Best Solution: Fitness {fitness(best_solution)}/{target}',
             fontsize=12, fontweight='bold')

# Evolution graph
axes[1].plot(best_fitness_history, color='steelblue', linewidth=2)
axes[1].axhline(target, color='red', linestyle='--', label=f'Target = {target}')
axes[1].set_xlabel('Generation'); axes[1].set_ylabel('Best Fitness')
axes[1].set_title('GA Evolution — fitness improves over generations',
                  fontsize=12, fontweight='bold')
axes[1].legend(); axes[1].grid(alpha=0.3)

plt.tight_layout(); plt.show()
```

## 🤖 Part 4 · Q-learning — a simple RL example in GridWorld

> **GridWorld**: an agent in a 4×4 grid. The goal is to reach **goal (3,3)** while avoiding the **pit (1,1)**.


```python
# Simple GridWorld + Q-learning
import numpy as np
import matplotlib.pyplot as plt

GRID_SIZE = 4
N_ACTIONS = 4   # 0=up, 1=down, 2=left, 3=right
GOAL = (3, 3)
PIT = (1, 1)

# Reward function
def get_reward(state):
    if state == GOAL: return 10
    if state == PIT:  return -10
    return -0.1  # small penalty for each step (encourages a shorter path)

def step(state, action):
    r, c = state
    if action == 0: r = max(r-1, 0)            # up
    elif action == 1: r = min(r+1, GRID_SIZE-1) # down
    elif action == 2: c = max(c-1, 0)           # left
    else: c = min(c+1, GRID_SIZE-1)             # right
    new_state = (r, c)
    return new_state, get_reward(new_state), new_state in [GOAL, PIT]

# Q-table: states × actions
Q = np.zeros((GRID_SIZE, GRID_SIZE, N_ACTIONS))

# Hyperparameters
alpha = 0.1     # learning rate
gamma = 0.9     # discount factor
epsilon = 1.0   # exploration probability (decays)
epsilon_decay = 0.995
episodes = 500

rewards_per_episode = []

np.random.seed(42)
for episode in range(episodes):
    state = (0, 0)
    total_reward = 0
    done = False
    while not done:
        # Epsilon-greedy action
        if np.random.random() < epsilon:
            action = np.random.randint(N_ACTIONS)  # explore
        else:
            action = np.argmax(Q[state[0], state[1]])  # exploit

        new_state, reward, done = step(state, action)

        # Q-learning update (Bellman equation)
        best_next = np.max(Q[new_state[0], new_state[1]])
        Q[state[0], state[1], action] += alpha * (reward + gamma * best_next - Q[state[0], state[1], action])

        state = new_state
        total_reward += reward

    rewards_per_episode.append(total_reward)
    epsilon = max(0.05, epsilon * epsilon_decay)

print(f"✅ Training completed after {episodes} episodes")
print(f"   Final epsilon: {epsilon:.3f}")
print(f"   Average reward over last 50 episodes: {np.mean(rewards_per_episode[-50:]):.2f}")
```

```python
# Visualisation: learnt policy + reward curve
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Policy (arrows in each cell)
ax = axes[0]
arrows = {0: '↑', 1: '↓', 2: '←', 3: '→'}
for r in range(GRID_SIZE):
    for c in range(GRID_SIZE):
        if (r, c) == GOAL:
            color = 'lightgreen'; label = 'GOAL'
        elif (r, c) == PIT:
            color = 'lightcoral'; label = 'PIT'
        else:
            color = 'white'
            best_action = np.argmax(Q[r, c])
            label = arrows[best_action]
        ax.add_patch(plt.Rectangle((c, GRID_SIZE-1-r), 1, 1, facecolor=color, edgecolor='black'))
        ax.text(c + 0.5, GRID_SIZE-1-r + 0.5, label, ha='center', va='center', fontsize=22)

ax.set_xlim(0, GRID_SIZE); ax.set_ylim(0, GRID_SIZE)
ax.set_aspect('equal'); ax.set_xticks([]); ax.set_yticks([])
ax.set_title(f'Learnt Policy (after {episodes} episodes)',
             fontsize=12, fontweight='bold')

# Reward curve
axes[1].plot(rewards_per_episode, alpha=0.3, color='steelblue')
# Moving average
window = 25
moving_avg = np.convolve(rewards_per_episode, np.ones(window)/window, mode='valid')
axes[1].plot(np.arange(window-1, len(rewards_per_episode)), moving_avg,
             color='darkblue', linewidth=2, label=f'Moving avg ({window} ep)')
axes[1].set_xlabel('Episode'); axes[1].set_ylabel('Total reward')
axes[1].set_title('Reward by episode (the agent is learning!)', fontweight='bold')
axes[1].legend(); axes[1].grid(alpha=0.3)

plt.tight_layout(); plt.show()
```

### 🎯 TRY IT #2 — What do you observe on the reward graph?

1. **First 100 episodes**: reward is negative or unstable → the agent is **exploring**
2. **After 200-300 episodes**: reward becomes consistently high → the agent has **learnt** the optimal policy (exploitation)
3. On the policy map: arrows **point towards GOAL** while avoiding PIT

> 💎 **This is the exploration → exploitation transition in action!** IB may ask *"Describe how an RL agent learns"* — this is a concrete example.

### 🏆 Top 10 tips for Section B Paper 1

1. **Read the case study TWICE** before writing. Underline key facts.
2. **Count the marks** — for a 6-mark question, write **6 clear ideas**.
3. **Quote the case study** — *"as the case study mentions..."* can earn support marks.
4. **Use syllabus terms** — supervised, recall, backpropagation, etc.
5. **Use concrete examples** in ethics — COMPAS, Amazon AI, Netflix, Uber.
6. **Include counter-arguments** in Discuss — without them, 4/6 is usually the ceiling.
7. **Conclusion** — *"In conclusion..."* is expected.
8. **No padding** — each sentence should earn a mark.
9. **Do not write code** — IB is not assessing Python syntax in these answers.
10. **Time management** — leave 30-40 minutes for Section B.

### 📝 Ready answer structures

#### "Identify and Describe" [3]:
> *"This is **[concept]** (1). It works by **[mechanism]** (1). For example, **[concrete case]** (1)."*

#### "Outline" [2]:
> *"**[Key idea sentence 1]** (1). **[Supporting detail / example sentence]** (1)."*

#### "Explain" [4]:
> *"**[Claim]** (1) because **[reason]** (1). This is supported by **[mechanism/evidence]** (1). Therefore **[conclusion]** (1)."*

#### "Discuss / Evaluate" [6]:
> *"**[Argument 1 + example]** (2). **However, [counter-argument + example]** (2). **In conclusion, [balanced recommendation]** (2)."*


### 🚨 Common Mistakes — DO NOT DO THESE!

| Mistake | Why it is bad | Better version |
|---|---|---|
| Confusing regression and classification | loses easy marks | **Continuous output = regression, categorical = classification** |
| Writing "machine learning" instead of a specific type | -1 mark for lack of precision | **"supervised classification with CNN"** |
| In ethics, writing only about privacy | too superficial | **Cover 3+ types: bias, accountability, privacy, transparency** |
| Discuss answer without a conclusion | -1-2 marks | **Always finish with "In conclusion..."** |
| Very long answers for 1-mark questions | wastes time | **1 mark = 1 sentence** |
| No concrete examples in ethics | can lose 2 marks | **Learn 4 case studies by heart** |
| Writing maths formulas without interpretation | only partial credit | **Explain what the formula means in words** |
