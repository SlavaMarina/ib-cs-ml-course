## 🧠 Part 1 · AI vs ML vs Deep Learning — the foundation

This is one of the most common mix-ups in the exam. Get it straight:

> **Artificial Intelligence (AI)** — a broad field focused on creating systems that can perform tasks normally requiring human intelligence, such as reasoning, perception, and decision-making.
>
> **Machine Learning (ML)** — a *subset* of AI. Algorithms learn **from data** without rules being explicitly programmed.
>
> **Deep Learning (DL)** — a *subset* of ML that uses **neural networks** with several layers of neurons.

### Visualisation — nested sets:


````python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(9, 7))

# Three nested circles
circle_ai = patches.Circle((0.5, 0.5), 0.45, color='#FFE5B4', ec='black', lw=2)
circle_ml = patches.Circle((0.5, 0.45), 0.30, color='#B4E5FF', ec='black', lw=2)
circle_dl = patches.Circle((0.5, 0.40), 0.15, color='#B4FFB4', ec='black', lw=2)

ax.add_patch(circle_ai)
ax.add_patch(circle_ml)
ax.add_patch(circle_dl)

ax.text(0.5, 0.90, 'Artificial Intelligence (AI)', ha='center', fontsize=14, fontweight='bold')
ax.text(0.5, 0.68, 'Machine Learning (ML)', ha='center', fontsize=12, fontweight='bold')
ax.text(0.5, 0.40, 'Deep\nLearning', ha='center', fontsize=11, fontweight='bold')

ax.text(0.5, 0.05, 'AI ⊃ ML ⊃ DL',
        ha='center', fontsize=13, style='italic', color='darkred')

ax.set_xlim(0, 1); ax.set_ylim(0, 1)
ax.set_aspect('equal'); ax.axis('off')
ax.set_title('Hierarchy: AI ⊃ ML ⊃ Deep Learning', fontsize=14, pad=15)
plt.tight_layout()
plt.show()
````

> 🔑 **Common mistake (Baumgarten, p. 4):**
> *"Deep learning is a subset of machine learning. Deep learning is not separate from machine learning, but rather is a specific approach within it."*
>
> In the exam, if the question asks *"Distinguish between machine learning and deep learning"*, do not write as if they are separate categories. DL is ML + neural networks.

### 💡 Narrow AI vs General AI

| | Narrow (Weak) AI | General (Strong) AI |
|---|---|---|
| What it does | one specific task | a wide range of tasks, like a human |
| Does it exist? | **Yes** — everything we see today | **No** — still theoretical |
| Examples | ChatGPT, face recognition, Netflix recommendations | — |

> ⚠️ **Exam secret #2:** if an examiner asks "is ChatGPT general AI?", the answer is **NO**. It is narrow AI because it is trained for a specific type of task: text.


## 🧩 Part 2 · Five learning paradigms (A4.1.1)

In the IB syllabus you need to know **5 paradigms**:

1. **Supervised learning** — learning from labelled examples
2. **Unsupervised learning** — learning without labels
3. **Reinforcement learning** — learning from rewards
4. **Deep learning** — learning with neural networks
5. **Transfer learning** — reusing learning from one task in a related task

---

### 🟢 1. Supervised Learning

> **Definition (learn this):** *Supervised learning is when a machine learning algorithm is provided with a data set of pairs of items, where the pair comprises an input value and the correct output response the algorithm should produce.*

Two subtypes:

| Type | Output | Example |
|---|---|---|
| **Regression** | a *continuous* numeric value | predicting house price ($), tomorrow's temperature (°C), or a mark (0-100) |
| **Classification** | a *discrete* category | spam / not spam · cat / muffin · cat / dog / parrot |

> 🚨 **Common mistake (Baumgarten):** "Confusing the goals of regression and classification".
> Regression = **continuous**, Classification = **categorical**.
> In the exam, *"Is house price prediction regression or classification?"* — **regression** because price is continuous.
> *"Is spam detection regression or classification?"* — **classification** because there are discrete categories.


````python
# Visualisation: Regression vs Classification
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# --- Regression
np.random.seed(42)
x = np.linspace(0, 10, 30)
y = 2*x + 5 + np.random.normal(0, 3, 30)
axes[0].scatter(x, y, color='steelblue', s=60, alpha=0.7)
axes[0].plot(x, 2*x + 5, 'r--', label='Regression line')
axes[0].set_xlabel('House area (tens of square metres)')
axes[0].set_ylabel('Price (hundreds of thousands of $)')
axes[0].set_title('REGRESSION — continuous output', fontsize=12, fontweight='bold')
axes[0].legend(); axes[0].grid(alpha=0.3)

# --- Classification
np.random.seed(0)
class_a = np.random.normal([3, 3], 0.7, (25, 2))
class_b = np.random.normal([7, 7], 0.7, (25, 2))
axes[1].scatter(class_a[:,0], class_a[:,1], color='orange', s=80, label='Class: Cat', edgecolor='k')
axes[1].scatter(class_b[:,0], class_b[:,1], color='purple', s=80, label='Class: Dog', edgecolor='k')
axes[1].plot([0, 10], [10, 0], 'g--', label='Decision boundary')
axes[1].set_xlabel('Feature 1')
axes[1].set_ylabel('Feature 2')
axes[1].set_title('CLASSIFICATION — discrete output', fontsize=12, fontweight='bold')
axes[1].legend(); axes[1].grid(alpha=0.3)

plt.tight_layout(); plt.show()
````

### 🟡 2. Unsupervised Learning

> **Definition:** *A method of machine learning where the data set does NOT include "answers" or expected outputs. The algorithm attempts to discover patterns on its own.*

**Two main subtypes (you need both for HL — A4.3.4, A4.3.5):**

| Subtype | What it does | Example |
|---|---|---|
| **Clustering** | groups similar objects | customer segmentation |
| **Association rule** | finds co-occurrences | "customers who bought coffee also bought biscuits" |

**Where it is used:**
- Social networks: analysing social groups without reading message contents, using contact frequency patterns
- Netflix, Spotify: recommendation systems
- Retail: market basket analysis

---

### 🔴 3. Reinforcement Learning

> **Definition:** *Machine learning by trial and error. The algorithm selects an action in an environment; the environment provides a reward; the algorithm refines its decision-making over time.*

**5 components (memorise them):** these are often asked in HL Section B.

| Component | Meaning |
|---|---|
| **Agent** | the learner or decision-maker |
| **Environment** | the world around the agent |
| **State** | a snapshot of the world at one moment |
| **Action** | what the agent does |
| **Reward** | feedback, positive or negative |

**Examples:** AlphaGo, autonomous cars, trading bots, training a robot to walk.

---

### 🔵 4. Deep Learning

> **Definition:** *Implies the use of a neural network within a machine learning algorithm.*

This is **not a separate paradigm** from ML. It is an implementation approach. You can say "deep supervised learning" for a neural network classifier, or "deep reinforcement learning" for an RL agent with a neural-network brain.

---

### 🟣 5. Transfer Learning

> **Definition:** *When a previously trained machine learning model is applied to a similar yet new situation. Goal: speed up training by reusing already learned features.*

**Example:** a model trained on ImageNet to recognise 1000 classes is fine-tuned to recognise dog breeds. This needs thousands of photos, not millions.

> ⚠️ **Common mistake (Baumgarten):** *"Transfer learning is not just about using a pre-trained model. It involves ADAPTING a model developed for one task to solve a related one."*
> Simply running a ready-made model is **not** transfer learning. Fine-tuning or adaptation is needed.


### 📊 Exam cheat sheet — comparison table

| Paradigm | Data | Goal | Algorithms | Example |
|---|---|---|---|---|
| **Supervised** | labelled data (input + label) | predict an output | KNN, Decision Tree, Linear Regression | diagnosing disease from symptoms |
| **Unsupervised** | unlabelled data | find structure | K-means, DBSCAN, Apriori | customer segmentation |
| **Reinforcement** | interaction with an environment | maximise reward | Q-learning, Policy Gradient | autonomous vehicle, AlphaGo |
| **Deep** | large or complex data | learn complex patterns | ANN, CNN, RNN | face recognition |
| **Transfer** | limited labelled data | reuse existing knowledge | fine-tuning | adapting GPT to a medical domain |

> 💎 **Exam secret #3:** For *"Identify the learning paradigm used for [scenario]"*, always look at the **data**:
> - Labels are present → supervised
> - No labels → unsupervised
> - Learns from rewards → reinforcement
> - Uses an existing model → transfer


## 🖥️ Part 3 · Hardware Requirements (A4.1.2)

For IB, you must know **6 processor types** and **5 computing platforms**.

### 🔧 Processor types

| Processor | What it is | When to use it | Key feature |
|---|---|---|---|
| **CPU** | general-purpose processor | basic ML, small models | flexible, but mostly sequential |
| **GPU** | hundreds/thousands of small cores | neural network training, parallel tasks | up to **10× faster than CPU** for ML |
| **TPU** | Google chip for tensors | deep learning, TensorFlow | efficient 16-bit tensor operations |
| **NPU** | found in Copilot AI laptops / Apple Silicon | edge inference on-device | 40+ TOPS for Copilot-style certification |
| **ASIC** | chip for one specific task | mass production (Apple A-series, Snapdragon) | maximum speed, but **not** reprogrammable |
| **FPGA** | reprogrammable chip | prototypes, high-frequency trading | balance of flexibility and speed |

> 🚨 **Common mistake (Baumgarten):** *"Confusing the differences between each of the processor types."*
> Key distinction:
> - **ASIC** — fixed, designed for ONE task → mass deployment
> - **FPGA** — reprogrammable → flexibility
> - **GPU** — parallel processing → neural networks
> - **TPU** — Google, tensors → large-scale DL
> - **NPU** — consumer edge devices → on-device AI


````python
# Comparison chart for processor types
import matplotlib.pyplot as plt
import numpy as np

processors = ['CPU', 'GPU', 'TPU', 'NPU', 'ASIC', 'FPGA']
flexibility = [10, 7, 5, 4, 1, 8]      # 1-10
ml_speed    = [2, 8, 9, 6, 10, 5]
energy_eff  = [3, 5, 7, 9, 10, 6]

x = np.arange(len(processors))
width = 0.27

fig, ax = plt.subplots(figsize=(11, 6))
ax.bar(x - width, flexibility, width, label='Flexibility', color='#4C72B0')
ax.bar(x,         ml_speed,    width, label='ML speed', color='#DD8452')
ax.bar(x + width, energy_eff,  width, label='Energy efficiency', color='#55A868')

ax.set_ylabel('Score (1-10)', fontsize=11)
ax.set_title('Processor comparison across 3 criteria', fontsize=13, fontweight='bold')
ax.set_xticks(x); ax.set_xticklabels(processors, fontsize=11)
ax.set_ylim(0, 11)
ax.legend(); ax.grid(axis='y', alpha=0.3)
plt.tight_layout(); plt.show()
````

### 🏭 Computing Platforms

| Platform | Best suited for | Advantages | Limitations |
|---|---|---|---|
| **Standard laptop** | study, prototypes | cheap, available | limited memory, slow |
| **Dedicated workstation** (with GPU) | serious projects | up to 10× faster than CPU | expensive, needs maintenance |
| **Edge devices** | on-device inference | privacy, low latency | requires suitable hardware |
| **Cloud platforms** (AWS, GCP, Azure) | training large models | scalable, no hardware purchase | provider dependence, monthly cost |
| **HPC centres** | scientific research | huge computing power | often requires membership or a grant |

---

### 🌐 Edge Computing — a separate exam topic ❗

> **Definition (for Define [1]):** *Edge computing refers to computing systems that perform data processing AT or NEAR the location where data is generated, rather than relying on centralized cloud resources.*

**Why use edge computing?**
- ⚡ **Low latency** — critical for autonomous vehicles and medical devices
- 🔒 **Privacy** — data does not leave the device
- 📡 **Offline operation** — a phone can still work without internet
- 💰 **Less data transfer** — saves bandwidth

> 💎 **Exam secret #4 (common in Paper 1 Section B):** if the question is about a "self-driving car" or "real-time" system, include edge computing + low latency. That can earn 1-2 marks.


## ⚙️ Part 4 · Training vs Inference — a critical distinction

| | **Training** | **Inference** |
|---|---|---|
| What happens | the model learns from data | the model is applied to new data |
| Resources needed | **very high** (GPU/TPU clusters) | **low** (can run on a phone) |
| Typical location | cloud, HPC, workstation | edge device, NPU, ASIC |
| Time | hours / days / weeks | milliseconds |

**Example:**
- *Training* a Face ID model: millions of photos, weeks on an Apple GPU cluster.
- *Inference* using Face ID on an iPhone: about 0.5 seconds on the phone's NPU.

> 💎 **Exam secret #5:** the training/inference distinction often gives easy marks.
> If the question asks *"Why does this app need GPU?"*, say whether it is for **training** (powerful GPU/TPU needed) or **inference** (NPU/CPU is often enough).


## 🎯 Part 5 · Top Tip from Baumgarten (p. 16)

Adapt this as a guide for choosing hardware in IB scenarios:

```
🔸 Large models, real-time processing?
   Yes → GPU (parallel processing)
   No  → TPU (batch processing, tensors)

🔸 Real-time inference, deployed on edge?
   Yes → NPU or ASIC (energy-optimised)
   No  → FPGA (flexibility) or ASIC (if the task will not change)

🔸 Will the model need updates?
   Yes → FPGA (reprogrammable)
   No  → ASIC or GPU

🔸 Is low cost more important than performance?
   Yes → older GPUs or cloud

🔸 Need fast scalability?
   Yes → Cloud GPUs / TPUs
```

> 📌 **In the exam:** structure your answer through these questions. The examiner can immediately see that you understand the trade-offs.
