## 🧠 Часть 1 · AI vs ML vs Deep Learning — фундамент

Это **самая частая путаница** на экзамене. Запоминаем чётко:

> **Artificial Intelligence (AI)** — широкое поле, цель которого — создание систем, способных выполнять задачи, обычно требующие человеческого интеллекта (рассуждение, восприятие, принятие решений).
>
> **Machine Learning (ML)** — *подмножество* AI. Алгоритмы учатся **из данных** без явного программирования правил.
>
> **Deep Learning (DL)** — *подмножество* ML, где используются **нейронные сети** (несколько слоёв нейронов).

### Визуализация — вложенные множества:

````python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(9, 7))

# Три вложенных круга
circle_ai = patches.Circle((0.5, 0.5), 0.45, color='#FFE5B4', ec='black', lw=2)
circle_ml = patches.Circle((0.5, 0.45), 0.30, color='#B4E5FF', ec='black', lw=2)
circle_dl = patches.Circle((0.5, 0.40), 0.15, color='#B4FFB4', ec='black', lw=2)

ax.add_patch(circle_ai)
ax.add_patch(circle_ml)
ax.add_patch(circle_dl)

ax.text(0.5, 0.90, 'Artificial Intelligence (AI)', ha='center', fontsize=14, fontweight='bold')
ax.text(0.5, 0.68, 'Machine Learning (ML)', ha='center', fontsize=12, fontweight='bold')
ax.text(0.5, 0.40, 'Deep\nLearning', ha='center', fontsize=11, fontweight='bold')

ax.text(0.5, 0.05, 'AI ⊃ ML ⊃ DL',
        ha='center', fontsize=13, style='italic', color='darkred')

ax.set_xlim(0, 1); ax.set_ylim(0, 1)
ax.set_aspect('equal'); ax.axis('off')
ax.set_title('Иерархия: AI ⊃ ML ⊃ Deep Learning', fontsize=14, pad=15)
plt.tight_layout()
plt.show()
````

## 🧩 Часть 2 · Пять парадигм обучения (A4.1.1)

В IB syllabus вы должны знать **5 парадигм**:

1. **Supervised learning** — обучение с учителем
2. **Unsupervised learning** — обучение без учителя
3. **Reinforcement learning** — обучение с подкреплением
4. **Deep learning** — глубокое обучение (с нейросетями)
5. **Transfer learning** — перенос обучения

---

### 🟢 1. Supervised Learning

> **Определение (выучить!)**: *Supervised learning is when a machine learning algorithm is provided with a data set of pairs of items, where the pair comprises an input value and the correct output response the algorithm should produce.*

Два подтипа:

| Тип | Выход | Пример |
|---|---|---|
| **Regression** | *непрерывное* числовое значение | Предсказать цену дома (₽), температуру завтра (°C), оценку (0–100) |
| **Classification** | *дискретная* категория | spam / not spam · собака / маффин · кошка / собака / попугай |

> 🚨 **Common mistake (Baumgarten):** "Confusing the goals of regression and classification".
> Regression = **continuous**, Classification = **categorical**.
> На экзамене на вопрос *"Is house price prediction regression or classification?"* — это **regression** (цена — непрерывная).
> *"Is spam detection regression or classification?"* — это **classification** (только 2 категории).


````python
# Визуализация Regression vs Classification
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# --- Regression
np.random.seed(42)
x = np.linspace(0, 10, 30)
y = 2*x + 5 + np.random.normal(0, 3, 30)
axes[0].scatter(x, y, color='steelblue', s=60, alpha=0.7)
axes[0].plot(x, 2*x + 5, 'r--', label='Линия регрессии')
axes[0].set_xlabel('Площадь дома (десятки м²)')
axes[0].set_ylabel('Цена (млн ₽)')
axes[0].set_title('REGRESSION — непрерывный выход', fontsize=12, fontweight='bold')
axes[0].legend(); axes[0].grid(alpha=0.3)

# --- Classification
np.random.seed(0)
class_a = np.random.normal([3, 3], 0.7, (25, 2))
class_b = np.random.normal([7, 7], 0.7, (25, 2))
axes[1].scatter(class_a[:,0], class_a[:,1], color='orange', s=80, label='Класс: Кошка', edgecolor='k')
axes[1].scatter(class_b[:,0], class_b[:,1], color='purple', s=80, label='Класс: Собака', edgecolor='k')
axes[1].plot([0, 10], [10, 0], 'g--', label='Граница решения')
axes[1].set_xlabel('Признак 1')
axes[1].set_ylabel('Признак 2')
axes[1].set_title('CLASSIFICATION — дискретный выход', fontsize=12, fontweight='bold')
axes[1].legend(); axes[1].grid(alpha=0.3)

plt.tight_layout(); plt.show()
````

### 🟡 2. Unsupervised Learning

> **Определение**: *A method of machine learning where the data set does NOT include "answers" or expected outputs. The algorithm attempts to discover patterns on its own.*

**Два главных подтипа (нужно знать оба для HL — A4.3.4, A4.3.5):**

| Подтип | Что делает | Пример |
|---|---|---|
| **Clustering** | группирует похожие объекты | Сегментация клиентов магазина |
| **Association rule** | ищет co-occurrences | "Кто купил пиво, тот купил подгузники" |

**Где используется:**
- Социальные сети (WhatsApp): анализ соц. групп даже без чтения сообщений (только по частоте контактов)
- Netflix, Spotify: рекомендательные системы
- Retail: анализ покупательских ассоциаций

---

### 🔴 3. Reinforcement Learning

> **Определение**: *Machine learning by trial and error. The algorithm selects an action in an environment; the environment provides a reward; the algorithm refines its decision-making over time.*

**5 компонентов (запомните!)** — это спрашивают в HL Section B:

| Компонент | Что это |
|---|---|
| **Agent** | тот, кто учится (бот, робот) |
| **Environment** | мир вокруг агента |
| **State** | снимок мира в момент времени |
| **Action** | что агент делает |
| **Reward** | обратная связь (+ или −) |

**Примеры:** AlphaGo, самоуправляемые авто, торговые боты, обучение робота ходить.

---

### 🔵 4. Deep Learning

> **Определение**: *Implies the use of a neural network within a machine learning algorithm.*

Это **НЕ отдельная парадигма** — это способ реализации. Можно сказать: "deep supervised learning" (нейросеть для классификации) или "deep reinforcement learning" (нейросеть как мозг RL-агента).

---

### 🟣 5. Transfer Learning

> **Определение**: *When a previously trained machine learning model is applied to a similar yet new situation. Goal: speed up training by reusing already learned features.*

**Пример:** модель, обученная распознавать 1000 классов на ImageNet, дообучается распознавать породы собак (нужны только тысячи фото, а не миллионы).

> ⚠️ **Common mistake (Baumgarten):** *"Transfer learning is not just about using a pre-trained model. It involves ADAPTING a model developed for one task to solve a related one."*
> Просто запустить готовую модель = **НЕ** transfer learning. Нужен fine-tuning.


### 📊 Шпаргалка для экзамена — сравнительная таблица

| Парадигма | Данные | Цель | Алгоритмы | Пример |
|---|---|---|---|---|
| **Supervised** | размеченные (input + label) | предсказать выход | KNN, Decision Tree, Linear Regression | диагностика болезней по симптомам |
| **Unsupervised** | без меток | найти структуру | K-means, DBSCAN, Apriori | сегментация клиентов |
| **Reinforcement** | через взаимодействие | максимизировать reward | Q-learning, Policy Gradient | беспилотный автомобиль, AlphaGo |
| **Deep** | большие, любые | сложные паттерны | ANN, CNN, RNN | распознавание лиц |
| **Transfer** | мало размеченных | переиспользовать знания | fine-tuning | дообучение GPT под медицинский домен |

> 💎 **СЕКРЕТ #3:** На вопрос *"Identify the learning paradigm used for [scenario]"* всегда смотрите на **данные**:
> - Есть метки → supervised
> - Нет меток → unsupervised
> - Учится по награде → reinforcement
> - Использует существующую модель → transfer


## 🖥️ Часть 3 · Hardware Requirements (A4.1.2)

В IB обязательно знать **6 типов процессоров** + **5 платформ**.

### 🔧 Типы процессоров

| Процессор | Что это | Когда использовать | Особенность |
|---|---|---|---|
| **CPU** | универсальный процессор | базовый ML, малые модели | гибкий, но последовательная обработка |
| **GPU** | сотни/тысячи мелких ядер | обучение нейросетей, параллельные задачи | до **10× быстрее CPU** для ML |
| **TPU** | разработан Google для тензоров | глубокое обучение, TensorFlow | 16-битные операции — высокая эффективность |
| **NPU** | в ноутбуках Copilot AI / Apple Silicon | edge inference (на устройстве) | min 40 TOPS для Copilot-сертификации |
| **ASIC** | под одну конкретную задачу | массовое производство (Apple A-chip, Snapdragon) | максимум скорости, но НЕ перепрограммируется |
| **FPGA** | можно перепрограммировать | прототипы, high-frequency trading | баланс гибкости и скорости |

> 🚨 **Common mistake (Baumgarten):** *"Confusing the differences between each of the processor types."*
> Запомните ключевое отличие:
> - **ASIC** — fixed, designed for ONE task → массовое использование
> - **FPGA** — reprogrammable → гибкость
> - **GPU** — parallel processing → нейросети
> - **TPU** — Google, tensors → large-scale DL
> - **NPU** — consumer edge devices → on-device AI


````python
# Сравнительная диаграмма процессоров
import matplotlib.pyplot as plt
import numpy as np

processors = ['CPU', 'GPU', 'TPU', 'NPU', 'ASIC', 'FPGA']
flexibility = [10, 7, 5, 4, 1, 8]      # 1-10
ml_speed    = [2, 8, 9, 6, 10, 5]
energy_eff  = [3, 5, 7, 9, 10, 6]

x = np.arange(len(processors))
width = 0.27

fig, ax = plt.subplots(figsize=(11, 6))
ax.bar(x - width, flexibility, width, label='Гибкость', color='#4C72B0')
ax.bar(x,         ml_speed,    width, label='Скорость ML', color='#DD8452')
ax.bar(x + width, energy_eff,  width, label='Энергоэффективность', color='#55A868')

ax.set_ylabel('Оценка (1–10)', fontsize=11)
ax.set_title('Сравнение процессоров по 3 критериям', fontsize=13, fontweight='bold')
ax.set_xticks(x); ax.set_xticklabels(processors, fontsize=11)
ax.set_ylim(0, 11)
ax.legend(); ax.grid(axis='y', alpha=0.3)
plt.tight_layout(); plt.show()
````

### 🏭 Computing Platforms

| Платформа | Когда подходит | Плюсы | Минусы |
|---|---|---|---|
| **Standard laptop** | учёба, прототипы | дёшево, доступно | мало памяти, медленно |
| **Dedicated workstation** (с GPU) | серьёзные проекты | до 10× быстрее CPU | дорого, нужно обслуживание |
| **Edge devices** | inference на устройстве | приватность, низкая задержка | нужно своё железо |
| **Cloud platforms** (AWS, GCP, Azure) | обучение больших моделей | масштабируемость, не нужно покупать железо | зависимость от провайдера, плата помесячно |
| **HPC centres** | научные исследования | огромная мощность | требует членства/гранта |

---

### 🌐 Edge Computing — отдельная тема ❗

> **Definition (для Define [1]):** *Edge computing refers to computing systems that perform data processing AT or NEAR the location where data is generated, rather than relying on centralized cloud resources.*

**Зачем нужен edge computing?**
- ⚡ **Низкая задержка (latency)** — критично для беспилотных авто, медицинских устройств
- 🔒 **Приватность** — данные не уходят с устройства
- 📡 **Работа без интернета** — телефон может работать оффлайн
- 💰 **Меньше передачи данных** — экономия трафика

> 💎 **СЕКРЕТ #4 (часто на Paper 1 Section B):** если вопрос про "self-driving car" или "real-time", в ответе **обязательно** упомяните edge computing + low latency. Это даёт +1–2 балла.


## ⚙️ Часть 4 · Training vs Inference — критическое различие

| | **Training (обучение)** | **Inference (использование)** |
|---|---|---|
| Что происходит | модель учится на данных | модель применяется к новым данным |
| Сколько ресурсов | **очень много** (GPU/TPU кластеры) | **мало** (можно на телефоне) |
| Где | cloud, HPC, workstation | edge device, NPU, ASIC |
| Время | часы / дни / недели | миллисекунды |

**Пример:**
- *Training* модели Face ID: миллионы фото, недели на GPU-кластере Apple.
- *Inference* модели Face ID на iPhone: 0.5 сек на NPU телефона.

> 💎 **СЕКРЕТ #5:** на экзамене различие training/inference часто даёт лёгкие баллы.
> Если вопрос *"Why does this app need GPU?"* — уточните, для **training** (нужен мощный GPU/TPU) или для **inference** (часто достаточно NPU/CPU).

