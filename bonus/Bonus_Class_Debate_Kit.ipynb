# 🎭 BONUS · Class Debate Kit — Ethics of AI/ML
## Complete kit for running 8 structured classroom debates

> ⏱ Duration: **45-60 minutes per debate** (you can run one per week)
> 🎯 Goal: deep engagement with AI ethics through **soft skills** + ready structure for **IB Discuss/Evaluate** questions
> 👥 Format: **3-5 people per side** + **2-3 judges** + a moderator

---

### 📋 What is in this kit

1. 🧑‍🏫 **Teacher's Guide** — how to organise a debate
2. 🎴 **Role Cards** — for all participants
3. ⚖️ **8 ready debates** with pro/con arguments, cases, and killer questions
4. 📊 **Scoring Rubric** (IB-aligned)
5. 🪞 **Reflection Templates** — after the debate

### 🎯 Why debates belong in an ML course

**Benefits for an IB student:**

- 📚 Ready pool of **case studies** for exam ethics questions
- 🎓 Practice structuring extended responses (counter-argument + conclusion)


## 🧑‍🏫 PART 1 · Teacher's Guide

### Pre-debate preparation (1 week before)

1. **Announce the topic** and divide the class into 3 groups:
   - 🟢 **PRO team** (3-5 students)
   - 🔴 **CON team** (3-5 students)
   - ⚖️ **JUDGES** (2-3 students) + 1 moderator

2. **Give out Role Cards** (see Part 2 below) and the **Debate Brief** for the chosen debate (Part 3).

3. **Home preparation**: each team discusses arguments and assigns roles (lead speaker, support, fact-checker).

---

### Debate day structure (45-60 min)

| Time | Stage | What happens |
|---|---|---|
| 0-3 min | **Intro** | Moderator reads the case + rules |
| 3-7 min | **Opening PRO** | Main argument from PRO team (4 min) |
| 7-11 min | **Opening CON** | Main argument from CON team (4 min) |
| 11-16 min | **Cross-examination** | CON asks PRO 3 killer questions (5 min) |
| 16-21 min | **Cross-examination** | PRO asks CON 3 killer questions (5 min) |
| 21-26 min | **Free debate** | Open discussion, moderated (5 min) |
| 26-30 min | **Closing PRO** | Final PRO speech with conclusion + mitigations (4 min) |
| 30-34 min | **Closing CON** | Final CON speech (4 min) |
| 34-44 min | **Judges deliberate** | Judges discuss and score using the rubric (10 min) |
| 44-50 min | **Verdict + reflection** | Verdict + insight discussion (5-10 min) |

---

### 3 formats to choose from

| Format | When to use it | Strengths |
|---|---|---|
| **Oxford-style** (as above) | Standard option for most groups | structured, IB-style |
| **Fishbowl** | Large classes (20+) | involves everyone |
| **Socratic seminar** | Philosophical topics | depth of reasoning |

**Fishbowl:** 4 students sit in the centre (2 PRO + 2 CON) and debate. Others sit around them and take notes. Every 5 minutes, **rotate**: one student leaves the centre and someone outside joins. Every student enters the fishbowl.

**Socratic seminar:** only **questions**, no statements. One student asks a question → another answers with a question → and so on. This teaches **questioning assumptions**.

---

### 📋 Teacher checklist

- [ ] Topic chosen **one week** in advance
- [ ] Teams formed (balance stronger/weaker students)
- [ ] **Role cards** printed / shared
- [ ] **Debate brief** with case given to **both** teams
- [ ] **Rubric** shown to judges before the start

> 💎 **Pro tip:** record debates (with permission), then share them with students. They often learn a lot from watching themselves.


## 🎴 PART 2 · Role Cards (for printing / handout)

> Each participant receives **one** card with their role and responsibilities.

---

### 🟢 PRO TEAM — Role Cards

#### 🎤 Lead Speaker (1 person)
**Your role:** OPENS the debate and CLOSES the debate.

**Responsibilities:**
- Opening speech (4 min): present the **main thesis** + 3 supporting arguments
- Closing speech (4 min): summarise + counter their critique + recommend mitigations

**What to prepare:**
- 1 central thesis ("X is justified because...")
- 3 strongest arguments (with cases!)
- Counter-arguments and how to refute them

> 💡 **Do not forget:** in the closing, mention **mitigation strategies**. This earns marks in IB-style answers.

---

#### 🛡 Support Speakers (1-2 persons)
**Your role:** SUPPORT the Lead Speaker with facts and cases.

**Responsibilities:**
- In cross-examination, answer questions with **facts** and **numbers**
- In free debate, add **extra case studies** and **examples**

**What to prepare:**
- 5+ concrete cases (COMPAS, Amazon AI, Netflix, Uber, GDPR fines)
- Numbers/statistics for backup
- "If they say X, we say Y" preparation

---

#### 🔍 Fact-Checker (1 person)
**Your role:** Track the factual accuracy of the opposition.

**Responsibilities:**
- Listen **critically** to CON arguments
- In free debate, point out **incorrect** claims ("Actually, the COMPAS algorithm was found...")
- Prepare **killer questions** for cross-examination

**What to prepare:**
- Deep knowledge of **3-4 cases** for verification
- A list of distortions the CON side might use

---

### 🔴 CON TEAM — same 3 roles, mirrored (Lead / Support / Fact-Checker)

---

### ⚖️ JUDGES (2-3 persons)

**Your role:** Evaluate the debate objectively and choose the winner.

**Responsibilities:**
- Listen WITHOUT bias (even if you have your own opinion)
- Score using the **Rubric** (see Part 4)
- During deliberation, discuss each rubric item
- In the verdict, announce the winner + give **constructive feedback** to both teams

**What to prepare:**
- Study the **Rubric** in advance
- Know the **basic** content of the debate

---

### 🎙 MODERATOR (1 person)

**Your role:** Manage time and order.

**Responsibilities:**
- Read the **brief case** at the start
- Announce each stage
- Watch the timer (do not be afraid to cut people off!)
- In free debate, alternate speakers (no shouting)
- Announce the judges' verdict at the end

> 💎 **Do not moderate the content.** Only time and order.

## 🎯 Debate #1: Autonomous Vehicles — Trolley Problem

### Scenario
> A self-driving car cannot brake in time. It can either: (a) continue straight and hit **3 pedestrians**; or (b) turn and crash into a wall, killing the **passenger inside**. Should the car be programmed to save **the majority** (utilitarian) or protect **the passenger at all costs**?

> ⚖️ **THE QUESTION:** *Should autonomous vehicles be programmed with utilitarian ethics (save the most lives)?*

### 🟢 PRO Arguments (For utilitarian)
1. **Math of saving lives**: saving 3 > saving 1 in any rational ethical calculation
2. **Public trust**: society will **not accept** a car that "kills strangers to save its passenger"
3. **Insurance + regulation**: governments already push for the "best outcome" → utilitarian logic wins
4. **Precedent**: emergency workers (doctors, firefighters) are trained for **utilitarian** decisions (triage)
5. **MIT Moral Machine** study (2018): most people prefer saving **more** people

### 🔴 CON Arguments (Against utilitarian)
1. **Market failure**: nobody will buy a car that may kill its owner. The industry collapses
2. **Slippery slope**: if a car "chooses" victims, what comes next? AI judge? AI jury?
3. **Property rights**: a bought car should **protect** its owner; that is its purpose
4. **Cultural variance**: utilitarianism is not universal. Other cultures may disagree
5. **Algorithmic discrimination risk**: utilitarian logic + AI bias could systematically sacrifice some groups

### 🔪 Killer Questions

**CON → PRO:** *"If utilitarian is right, would you accept that your car kills you to save 3 strangers? Would your parents agree to put you in such a car?"*

**PRO → CON:** *"If the owner is always protected, what about a school bus full of children? Should one driver's life always outweigh 30 children?"*

### 📚 Real-world cases to reference
- **Tesla 2018 crash** (driver fatality, no pedestrian)
- **Uber 2018 fatal crash** (pedestrian, no driver action)
- **Mercedes 2016 statement**: said cars would prioritise passengers (huge backlash)
- **MIT Moral Machine experiment**: 40M responses from 233 countries → cultural variance

## 🎯 Debate #2: AI Judges — Should AI Replace Human Judges?

### Scenario
> Courts are overloaded, and human judges can be biased. A startup proposes an **AI judge** for civil cases (gradually, not criminal cases). The AI is trained on 100,000 past cases. Pilot studies show AI consistency is **25% higher**, and decisions are **10× faster**.

> ⚖️ **THE QUESTION:** *Should AI gradually replace human judges in civil cases?*

### 🟢 PRO
1. **Consistency**: AI is not affected by mood or hunger (study: hungry judges denied parole 65% vs 35% after lunch — Israeli courts 2011)
2. **Speed**: 10× faster → better access to justice for poorer people waiting for years
3. **Cost**: judges are expensive; AI scales cheaply
4. **Removes implicit bias**: AI can be audited and adjusted in ways human intuition cannot
5. **Audit trail**: every AI decision can be checked; a human judge can be a black box

### 🔴 CON
1. **COMPAS warning**: AI can **amplify** racial bias
2. **No empathy**: civil cases often require **human understanding** (divorce, custody)
3. **Accountability gap**: who is sued if the AI is wrong? Company? Startup? Government?
4. **Training data is past data**: if past decisions were biased, AI **inherits** them (Amazon 2018 lesson)
5. **Democratic legitimacy**: judges are part of democracy. AI is not elected by society

### 🔪 Killer Questions
**CON → PRO:** *"COMPAS already showed that AI can inherit bias. What guarantees can you give that your AI judge will be fair?"*

**PRO → CON:** *"Human judges have 30%+ inconsistency rates. Why is your standard 'perfect AI or no AI' when humans are clearly imperfect?"*

### 📚 Cases
- **COMPAS** (ProPublica 2016)
- **Estonia AI judges** pilot 2019 (small claims <€7000)
- **China's AI judges** (used in some lower courts)
- **Loomis v. Wisconsin** (2016): court allowed COMPAS use in sentencing


## 🎯 Debate #3: Facial Recognition in Public — Ban or Allow?

### Scenario
> A city wants to deploy widespread facial recognition (FR) in public spaces to prevent crime. Pilot data from London (Met Police 2019) says FR helped find 8 wanted criminals out of 142,000 scanned faces.

> ⚖️ **THE QUESTION:** *Should mass facial recognition in public spaces be banned?*

### 🟢 PRO Ban
1. **Mass surveillance**: every scan is an invasion of privacy without consent
2. **Bias**: NIST 2019 study found much higher error rates for darker-skinned women
3. **Chilling effect**: people change behaviour when they know they can be identified
4. **No due process**: false match → arrest without proper review, as in **Robert Williams (Detroit 2020)**
5. **Misuse**: Clearview AI scraped billions of photos without consent and sold access to police

### 🔴 CON Ban (Allow with controls)
1. **Public safety**: missing children and terrorism prevention are real use cases
2. **CCTV is already legal** — FR is a more effective version. Where is the boundary?
3. **Strict regulations** are possible: warrant requirement, audit logs, bias audits
4. **Voluntary opt-out** zones for concerned citizens
5. **Catch-22**: complete ban → failure to prevent a serious attack → public outcry

### 🔪 Killer Questions
**PRO → CON:** *"You support warrants. But mass FR scans EVERYONE without a warrant. How is that constitutional?"*

**CON → PRO:** *"Robert Williams was one mistake among many correct matches. Should we ban all medicine because some patients die from misdiagnosis?"*

### 📚 Cases
- **Robert Williams Detroit 2020** (wrongful arrest)
- **NIST 2019 bias study** (Joy Buolamwini)
- **San Francisco / Boston / Portland** banned FR for police (2019-2020)
- **EU AI Act** classifies many FR uses as high-risk or banned
- **Clearview AI** scandal (2020)
- **London Metropolitan Police** trials

## 🎯 Debate #4: Social Media Algorithms — Should Platforms Be Liable?

### Scenario
> Facebook recommendation algorithms can create echo chambers and political polarisation. After Jan 6 2021 (Capitol riot), pressure grew to make platforms **legally liable** for amplified misinformation.

> ⚖️ **THE QUESTION:** *Should social media platforms be legally liable for harm caused by their recommendation algorithms?*

### 🟢 PRO Liability
1. **They CHOOSE** what to amplify; it is not a neutral pipe
2. **Profit from engagement** = profit from outrage = profit from division
3. **Section 230** (US law) is outdated; it was written before ML recommendation algorithms existed
4. **Other industries are liable**: TV networks and newspapers can be sued for harm
5. **Facebook's own research** (Frances Haugen leaks 2021): they **knew** Instagram harms teen mental health

### 🔴 CON Liability
1. **Free speech**: liability → over-censorship → silencing legitimate voices
2. **Impossible to police 5 billion users** — algorithms are necessary
3. **User responsibility**: people **choose** what to watch; the algorithm shows options
4. **Innovation chilling effect**: small startups may not survive this level of regulation
5. **Slippery slope**: today fake news, tomorrow political dissent is silenced

### 🔪 Killer Questions
**PRO → CON:** *"If Facebook's algorithm knowingly amplifies content that harms teenagers, why should it be immune when other manufacturers can be sued for harm?"*

**CON → PRO:** *"Who decides what counts as misinformation? Government? Facebook? Where is the line between fact-checking and censorship?"*

### 📚 Cases
- **Frances Haugen Facebook Files** (2021)
- **Section 230 US debate** (Communications Decency Act)
- **EU Digital Services Act** (2022) — does hold platforms accountable
- **YouTube radicalisation** studies
- **Myanmar genocide** (Facebook acknowledged its role, 2018)


## 🎯 Debate #5: AI-Generated Art — Theft or Tool?

### Scenario
> Generative AI (Midjourney, DALL-E, Stable Diffusion) was trained on billions of internet images **without artist consent**. Real artists are losing commissions to AI tools. Lawsuits have been filed (Getty vs Stability AI, 2023).

> ⚖️ **THE QUESTION:** *Is training generative AI on copyrighted images without consent fair use?*

### 🟢 PRO It's Fair Use
1. **Transformation**: AI creates new content, not copies (legal fair use criterion)
2. **Photographers learn** by looking at thousands of photos — same principle
3. **Democratisation**: people who cannot afford artists can now make good images
4. **Innovation**: blocking AI would prevent the next innovation, like colour photography changed art
5. **Open-source**: Stable Diffusion is free → benefits everyone, not only rich companies

### 🔴 CON Not Fair Use (Theft)
1. **No consent**: Getty's 12 million photos were used without permission
2. **Direct competition**: AI images compete with artists' income — not just transformation
3. **Style replication**: prompts like "in the style of [living artist]" → unfair appropriation
4. **Scale matters**: a human studies 1,000 images; AI processes billions → quantitatively different
5. **Watermarks found**: Stable Diffusion outputs sometimes show Getty-style watermarks → memorisation risk

### 🔪 Killer Questions
**CON → PRO:** *"If I scan all your essays and create AI that writes 'in your style', is that fair use? Why is artist work different?"*

**PRO → CON:** *"If we banned AI from learning from copyrighted work, would we also ban art schools from showing students copyrighted paintings? Where is the line?"*

### 📚 Cases
- **Getty Images vs Stability AI** (2023 lawsuit)
- **Sarah Andersen vs DeviantArt** (artists' class action 2023)
- **EU AI Act** requires disclosure of training data
- **Japan**: AI training legal regardless of consent (controversial)
- **Adobe Firefly**: trained only on licensed images (alternative model)

## 🎯 Debate #6: AI in Hiring — Efficiency or Discrimination?

### Scenario
> A corporation receives 5,000 CVs for one vacancy. AI performs the first filter in 5 seconds (vs 30 hours for human HR). But Amazon already shut down its recruiter tool in 2018 because of discrimination.

> ⚖️ **THE QUESTION:** *Should companies be allowed to use AI for resume screening?*

### 🟢 PRO Allow
1. **Scale problem**: 5,000 resumes × 30 min = 2,500 hours. Impossible without AI
2. **Less human bias**: human recruiters discriminate by name, address, and gender unconsciously (Harvard 2003 study)
3. **Better matches**: AI can find patterns humans miss (skills correlation, etc.)
4. **Audit trail**: AI decisions can be logged + reviewed; human decisions often cannot
5. **Already happens**: many large companies use ATS (Applicant Tracking Systems)

### 🔴 CON Ban (or Strict regulation)
1. **Amazon 2018 lesson**: AI **discriminated against women** because it was trained on male-dominated past hires
2. **Bias multiplier**: ML does not automatically reduce bias; it can **amplify** historical bias
3. **Black box**: rejected candidates **cannot appeal** if they do not know why they were rejected
4. **Proxy discrimination**: AI may use postcode or school as proxies for race/class
5. **EU AI Act** classifies hiring AI as high-risk → strict requirements are coming

### 🔪 Killer Questions
**PRO → CON:** *"If we ban AI hiring, we go back to humans rejecting resumes based on gut feeling. Is that better?"*

**CON → PRO:** *"Amazon ran its AI for 4 years before noticing the bias. How can you guarantee yours will not have similar undetected bias?"*

### 📚 Cases
- **Amazon 2018** (scrapped AI recruiter)
- **HireVue** (AI video interview controversy)
- **Workday class action** (2024, alleged age discrimination)
- **EEOC guidance** (2023, on AI in hiring)
- **EU AI Act** high-risk classification

## 🎯 Debate #7: AI in Schools — Cheating Tool or Learning Aid?

### Scenario
> ChatGPT was released in Nov 2022 and schools panicked. Some banned it; others embraced it. Students use it for homework and essays. Teachers cannot reliably detect AI-written text.

> ⚖️ **THE QUESTION:** *Should schools allow students to use AI tools like ChatGPT?*

### 🟢 PRO Allow
1. **Real-world preparation**: students will use AI in their careers anyway
2. **Personalised tutor**: wealthy students have tutors; AI can be a tutor for everyone (equity)
3. **Higher-order thinking**: assignments can become "critique this AI essay" → better learning
4. **Calculators precedent**: 1970s panic about calculators. Maths education adapted; it did not collapse
5. **Productivity**: students focus on **understanding**, not mechanical writing

### 🔴 CON Ban / Strict regulation
1. **Erodes critical thinking**: if AI does the thinking, the skill is not practised
2. **Cheating ≠ learning**: getting the answer is not the same as understanding it
3. **Inequity**: wealthy students get **better AI** (paid GPT-4, Claude Opus) → larger gap
4. **Devalues education**: degrees mean less if AI did the work
5. **Foundation matters**: students must learn basics **without** AI before using AI well

### 🔪 Killer Questions
**PRO → CON:** *"You ban ChatGPT but allow Google. Where is the line? Both are tools students could use to cheat."*

**CON → PRO:** *"If AI does the hard parts, students may never learn from struggle. How do you prevent that?"*

### 📚 Cases
- **NYC schools ChatGPT ban** (Jan 2023) → **reversed** May 2023
- **Australian universities**: hybrid approach (use AI but cite it)
- **Khan Academy Khanmigo**: school-approved AI tutor
- **Stanford pilots** with AI-assisted learning

## 🎯 Debate #8: Universal Basic Income — Should AI Profits Fund UBI?

### Scenario
> ML/AI may eliminate jobs (truck drivers, paralegals, radiologists eventually). Andrew Yang's 2020 presidential campaign popularised **UBI** ($1000/month). Should AI companies pay an extra tax to fund UBI?

> ⚖️ **THE QUESTION:** *Should AI companies pay a special tax to fund Universal Basic Income for displaced workers?*

### 🟢 PRO UBI Tax
1. **Magnitude**: 47% of US jobs at risk (Oxford 2013 study); unprecedented disruption
2. **Concentration**: 5 companies (Google, Meta, Microsoft, OpenAI, Amazon) capture much of AI profit
3. **Social contract**: society **subsidised AI research** (universities, government grants) — fair to recoup
4. **Stability**: mass unemployment → political extremism (history lesson: 1930s Germany)
5. **Pilot evidence**: Finland 2017-2018 UBI experiment showed positive mental health effects

### 🔴 CON No special tax
1. **Innovation killer**: special taxes → companies leave for friendlier jurisdictions
2. **What counts as "AI"?**: almost every software system has some AI. Hard to define
3. **History of automation**: agricultural revolution displaced many farmers; new jobs emerged
4. **Disincentive to work**: UBI experiments are mixed; some show reduced labour participation
5. **Better solutions**: education + retraining programmes > UBI

### 🔪 Killer Questions
**PRO → CON:** *"The Industrial Revolution displaced workers over centuries. AI may displace them in years. How will society absorb this without intervention?"*

**CON → PRO:** *"If you tax AI heavily, AI development may move abroad. Are you comfortable losing leadership in a strategic technology?"*

### 📚 Cases
- **Andrew Yang 2020** Freedom Dividend ($1000/month)
- **Finland UBI experiment** 2017-2018
- **Sam Altman (OpenAI) UBI advocacy**
- **Oxford study** (Frey & Osborne, 2013)
- **Stockton SEED program** (California UBI pilot)


## 📊 PART 4 · Scoring Rubric (IB-Aligned)

> Judges assess **each team** using this rubric. Total = 30 marks.

### Rubric (for printing for judges)

| Criterion | 4 (excellent) | 3 (good) | 2 (acceptable) | 1 (weak) |
|---|---|---|---|---|
| **Argument quality** (×3) | Multiple strong, well-reasoned arguments with logic chains | 2-3 solid arguments | 1-2 basic arguments | Vague or contradictory |
| **Evidence & cases** (×3) | 4+ specific real cases with numbers/dates | 2-3 relevant cases | 1 case | No specific evidence |
| **Counter-argument handling** (×2) | Anticipates and refutes opposition skilfully | Addresses most opposing points | Addresses some | Ignores opposition |
| **Use of IB terminology** (×2) | Uses 5+ syllabus terms naturally (bias, accountability, GDPR, etc.) | 3-4 terms used | 1-2 terms | None |
| **Speaking & rhetoric** (×2) | Clear, persuasive, varied tone | Clear, organised | Mostly clear | Difficult to follow |
| **Teamwork** (×1) | Seamless coordination, equal participation | Good coordination | One person dominates | Disorganised |
| **Mitigation proposals** (×2) | Concrete, realistic mitigations proposed | Some mitigations | Vague mitigations | None |
| **TOTAL** | **/30** | | | |

---

### 🏆 Scoring guide for judges

| Total | Verdict |
|---|---|
| **24-30** | Excellent debate — victory |
| **18-23** | Strong performance |
| **12-17** | Acceptable |
| **0-11** | Needs significant improvement |

> 💎 **Judges' tip:** do not score based on **your own** opinion. Score the **quality of argumentation**.
> You can disagree with the topic but still recognise that the other team argued better.

### 🎁 Bonus points (for the final class leaderboard)

- 🥇 Winner team: +3 marks each
- 🥈 Losing team: +1 mark for participation
- 🌟 "Outstanding speaker" (class vote): +2 marks
- 🔍 "Best killer question": +1 mark


## 🔗 PART 6 · Connection to IB IA

> Debates are a **gold mine** for IA. Here is how to use them.

### What IA can grow from these debates?

1. **Ethical Analysis IA**: take **one** debate → deepen it → interview stakeholders
2. **System Design with Ethics Review**: use **insights** from the debate for a **dedicated ethics section** in your IA
3. **Mitigation Strategies Catalogue**: collect **all** mitigation strategies from the 8 debates → create a **handbook** for ML developers

### IA structure with a debate-based opening

```
1. INTRODUCTION
   - State the problem from the debate
   - Context: "This topic was debated in our class on [date]"

2. STAKEHOLDER ANALYSIS
   - 3 interest groups (from debate teams)

3. TECHNICAL SOLUTION
   - Your ML approach to the problem

4. ETHICAL EVALUATION
   - PRO arguments (cite debate)
   - CON arguments (cite debate)
   - Mitigation matrix

5. CONCLUSION
   - Balanced recommendation
   - Future research needed
```

> 💎 **Pro tip:** moderators and judges often write the best IAs because they **listened to both sides** without bias. If you are a judge, use your notes.
