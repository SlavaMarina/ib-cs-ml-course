## 🧑‍🏫 PART 1 · Teacher's Guide

### Pre-debate preparation (за 1 неделю)

1. **Объявите тему** + поделите класс на 3 группы:
   - 🟢 **PRO team** (3-5 студентов)
   - 🔴 **CON team** (3-5 студентов)
   - ⚖️ **JUDGES** (2-3 студента) + 1 модератор

2. **Раздайте Role Cards** (см. Part 2 ниже) и **Debate Brief** для конкретного дебата (Part 3).

3. **Подготовка дома**: каждая команда обсуждает аргументы, делит роли (lead speaker, support, fact-checker).

---

### Debate day structure (45-60 мин)

| Время | Этап | Что происходит |
|---|---|---|
| 0-3 мин | **Intro** | Модератор зачитывает кейс + правила |
| 3-7 мин | **Opening PRO** | Главный аргумент PRO команды (4 мин) |
| 7-11 мин | **Opening CON** | Главный аргумент CON команды (4 мин) |
| 11-16 мин | **Cross-examination** | CON задаёт PRO 3 killer questions (5 мин) |
| 16-21 мин | **Cross-examination** | PRO задаёт CON 3 killer questions (5 мин) |
| 21-26 мин | **Free debate** | Открытое обсуждение, модератор управляет (5 мин) |
| 26-30 мин | **Closing PRO** | Финальная речь PRO с conclusion + mitigations (4 мин) |
| 30-34 мин | **Closing CON** | Финальная речь CON (4 мин) |
| 34-44 мин | **Judges deliberate** | Судьи обсуждают, оценивают по rubric (10 мин) |
| 44-50 мин | **Verdict + reflection** | Объявление + обсуждение insights (5-10 мин) |

---

### 3 формата на выбор

| Формат | Когда использовать | Плюсы |
|---|---|---|
| **Oxford-style** (как выше) | Стандартный, для всех групп | структурировано, IB-style |
| **Fishbowl** | Большие классы (20+) | вовлекает всех |
| **Socratic seminar** | Тематика философская | глубина рассуждений |

**Fishbowl:** 4 студента в центре (2 PRO + 2 CON) дебатируют. Остальные сидят кругом и пишут заметки. Через 5 мин — **смена** (один в центре уходит, кто-то снаружи приходит). Каждый студент проходит через "fishbowl".

**Socratic seminar:** только **вопросы**, никаких утверждений. Один задаёт вопрос → другой отвечает вопросом → и так далее. Учит **questioning assumptions**.

### 🟢 PRO TEAM — Role Cards

#### 🎤 Lead Speaker (1 person)
**Ваша роль:** ОТКРЫВАЕТ дебат + ЗАКРЫВАЕТ дебат.

**Обязанности:**
- Opening speech (4 мин): представить **главный thesis** + 3 поддерживающих аргумента
- Closing speech (4 мин): summarize + counter их critique + **rec mitigations**

**Что подготовить:**
- 1 центральный thesis ("X is justified because...")
- 3 strongest arguments (с кейсами!)
- Анти-аргументы и как их опровергать

> 💡 **Не забудьте:** в closing **обязательно** упомянуть **mitigation strategies**. Это даёт +balls на IB.

---

#### 🛡 Support Speakers (1-2 persons)
**Ваша роль:** ПОДКРЕПЛЯЕТ Lead Speaker фактами и кейсами.

**Обязанности:**
- В cross-examination — отвечать на questions с **фактами** и **числами**
- В free debate — приводить **доп. case studies** и **примеры**

**Что подготовить:**
- 5+ конкретных случаев (COMPAS, Amazon AI, Netflix, Tesla, GDPR fines)
- Numbers/statistics для backup
- "If they say X, we say Y" подготовка

---

#### 🔍 Fact-Checker (1 person)
**Ваша роль:** Следит за фактологией оппонентов.

**Обязанности:**
- Слушать **критически** аргументы CON
- В free debate — указывать **некорректные** утверждения ("Actually, the COMPAS algorithm was found...")
- Готовить **killer questions** для cross-examination

**Что подготовить:**
- Глубокое знание **3-4 кейсов** для verification
- Список distortions, которые CON могут использовать

---

### 🔴 CON TEAM — те же 3 роли, зеркально (Lead / Support / Fact-Checker)

### ⚖️ JUDGES (2-3 persons)

**Обязанности:**
- Слушать БЕЗ предвзятости (даже если у вас есть мнение)
- Оценивать по **Rubric** (см. Part 4)

- В verdict — объявить winner + **constructive feedback** обеим командам

**Что подготовить:**
- Изучить **Rubric** заранее

### 🎙 MODERATOR (1 person)

**Ваша роль:** Управляет временем и порядком.

**Обязанности:**
- Зачитать **brief case** в начале
- Объявлять каждый этап
- Следить за таймером (НЕ бояться обрывать!)
- В free debate — давать **слово** альтернативно (no shouting)
- Объявлять verdict судей в конце

## 🎯 Debate #1: Autonomous Vehicles — Trolley Problem

### Scenario
> Беспилотный автомобиль не успевает затормозить. Он может: (a) ехать прямо и сбить **3 пешеходов**; (b) свернуть и врезаться в стену, убив **пассажира внутри**. Должен ли автомобиль быть запрограммирован спасать **большинство** (utilitarian) или защищать **пассажира любой ценой**?

> ⚖️ **THE QUESTION:** *Should autonomous vehicles be programmed with utilitarian ethics (save the most lives)?*

### 🟢 PRO Arguments (For utilitarian)
1. **Math of saving lives**: спасти 3 > 1 по любой rational ethics
2. **Public trust**: общество **не примет** car, which "kills strangers to save its passenger"
3. **Insurance + regulation**: государство **уже** требует прав "best outcome" → utilitarian wins
4. **Precedent**: emergency workers (doctors, firefighters) тренированы для **utilitarian** decisions (triage)
5. **MIT Moral Machine** study (2018): большинство людей предпочитает спасать **больше** людей

### 🔴 CON Arguments (Against utilitarian)
1. **Market failure**: никто **не купит** car, которая может убить владельца. Industry collapses
2. **Slippery slope**: если car "выбирает" жертв — куда дальше? AI-judge? AI-jury?
3. **Property rights**: купленный автомобиль должен **защищать** владельца — это его purpose
4. **Cultural variance**: utilitarianism — Западная философия. Другие культуры (Japan, China) могут не согласиться
5. **Algorithmic discrimination risk**: utilitarian + AI bias → системно жертвует определённые группы

### 🔪 Killer Questions

**CON → PRO:** *"If utilitarian is right, would you accept that your car kills you to save 3 strangers? Would your parents agree to put you in such a car?"*

**PRO → CON:** *"If owner is always protected, what about a school bus full of children? Should one driver's life always outweigh 30 children?"*

### 📚 Real-world cases to reference
- **Tesla 2018 Arizona crash** (driver, no pedestrian)
- **Uber 2018 fatal crash** (pedestrian, no driver action)
- **Mercedes 2016 statement**: officially said cars would prioritize passenger (huge backlash)
- **MIT Moral Machine experiment**: 40M responses from 233 countries → cultural variance

## 🎯 Debate #2: AI Judges — Should AI Replace Human Judges?

### Scenario
> Перегруженные суды + bias людей-судей. Стартап предлагает **AI judge** для civil cases (gradually, не criminal). AI обучен на 100,000 past cases. По pilot studies, AI consistency **выше** на 25%, скорость решений **в 10 раз** быстрее.

> ⚖️ **THE QUESTION:** *Should AI gradually replace human judges in civil cases?*

### 🟢 PRO
1. **Consistency**: AI не зависит от настроения, голода (study: hungry judges denied parole 65% vs 35% after lunch — Israeli courts 2011)
2. **Speed**: 10× быстрее → доступ к правосудию для бедных, которые ждут годами
3. **Cost**: судьи дорогие; AI масштабируется бесплатно
4. **Removes implicit bias**: study Stanford 2017 — AI отказывает меньше людей с "не-белыми" именами
5. **Audit trail**: можно проверить **каждое** решение AI; человек-судья — black box

### 🔴 CON
1. **COMPAS catastrophe**: уже доказано — AI **усиливает** racial bias
2. **No empathy**: civil cases часто требуют **human understanding** (divorce, custody)
3. **Accountability gap**: кого судить, если AI ошибся? Compaq? Startup? Government?
4. **Training data is past data**: если в past были biased решения — AI **наследует** их (Amazon 2018 lesson)
5. **Democratic legitimacy**: судьи — часть democracy. AI не выбирается обществом

### 🔪 Killer Questions
**CON → PRO:** *"COMPAS уже доказал что AI наследует bias. Какие вы можете дать гарантии, что ваш AI judge будет fair?"*

**PRO → CON:** *"Human judges have 30%+ inconsistency rates. Why is your standard 'perfect AI or no AI' when humans are clearly imperfect?"*

### 📚 Cases
- **COMPAS** (ProPublica 2016)
- **Estonia AI judges** pilot 2019 (small claims <€7000)
- **China's AI judges** (used in some lower courts)
- **Loomis v. Wisconsin** (2016): supreme court allowed COMPAS use in sentencing


## 🎯 Debate #3: Facial Recognition in Public — Ban or Allow?

### Scenario
> Город хочет deploy widespread facial recognition (FR) в public spaces для предотвращения crime. По pilot data в London (Met Police 2019) — FR помог найти 8 wanted criminals из 142,000 сканированных лиц.

> ⚖️ **THE QUESTION:** *Should mass facial recognition in public spaces be banned?*

### 🟢 PRO Ban
1. **Mass surveillance**: каждый scan = invasion of privacy (без consent)
2. **Bias**: NIST 2019 study — FR в 100× менее accurate для darker-skinned women
3. **Chilling effect**: люди меняют поведение, зная что их identify (бывшая Stasi East Germany)
4. **No due process**: false match → arrest без рассмотрения, как в случае **Robert Williams (Detroit 2020)**
5. **Misuse**: Clearview AI scraped 3 billion фото без consent → продаёт полиции

### 🔴 CON Ban (Allow with controls)
1. **Public safety**: missing children, terrorism prevention — реальные wins
2. **CCTV уже legal** — FR просто более эффективная версия. Где граница?
3. **Strict regulations** возможны: warrant requirement, audit logs, bias audits
4. **Voluntary opt-out** zones для concerned citizens
5. **Catch-22**: complete ban → fail to prevent next mass shooting → public outcry

### 🔪 Killer Questions
**Pro → CON:** *"You support warrants. But mass FR scans EVERYONE without warrant. How is that constitutional?"*

**CON → PRO:** *"Robert Williams was 1 mistake in millions of correct matches. Should we ban all medicine because some patients die from misdiagnosis?"*

### 📚 Cases
- **Robert Williams Detroit 2020** (wrongful arrest)
- **NIST 2019 bias study** (Joy Buolamwini)
- **San Francisco / Boston / Portland** banned FR for police (2019-2020)
- **EU AI Act** classifies most FR as "high-risk" or banned
- **Clearview AI** scandal (2020)
- **London Metropolitan Police** trials

## 🎯 Debate #4: Social Media Algorithms — Should Platforms Be Liable?

### Scenario
> Facebook recommendation algorithm создаёт echo chambers → political polarization. После Jan 6 2021 (Capitol riot) — растёт давление сделать platforms **legally liable** for amplified misinformation.

> ⚖️ **THE QUESTION:** *Should social media platforms be legally liable for harm caused by their recommendation algorithms?*

### 🟢 PRO Liability
1. **They CHOOSE** what to amplify (it's not neutral pipe)
2. **Profit from engagement** = profit from outrage = profit from division
3. **Section 230** (US law) outdated — written before ML algorithms existed
4. **Other industries are liable**: TV networks, newspapers can be sued for harm
5. **Facebook own research** (Frances Haugen leaks 2021): they **knew** Instagram harms teen mental health

### 🔴 CON Liability
1. **Free speech**: making platforms liable → over-censorship → silencing legitimate voices
2. **Impossible to police 5 billion users** — algorithms necessary
3. **User responsibility**: люди **выбирают** что смотреть; algorithm just shows options
4. **Innovation chilling effect**: small startups не выживут с такой regulation
5. **Slippery slope**: today fake news, tomorrow — political dissent silenced

### 🔪 Killer Questions
**PRO → CON:** *"If Facebook's algorithm KNOWINGLY amplifies content that drives teen suicide, why should they be immune when a knife manufacturer can be sued?"*

**CON → PRO:** *"Who decides what's 'misinformation'? Government? Facebook? Where's the line between fact-checking and censorship?"*

### 📚 Cases
- **Frances Haugen Facebook Files** (2021)
- **Section 230 US debate** (Communications Decency Act)
- **EU Digital Services Act** (2022) — does hold platforms accountable
- **YouTube radicalization** studies
- **Myanmar genocide** (Facebook acknowledged its role, 2018)


## 🎯 Debate #5: AI-Generated Art — Theft or Tool?

### Scenario
> Generative AI (Midjourney, DALL-E, Stable Diffusion) trained на miliards of internet images **without artist consent**. Real artists losing commissions to AI tools. Lawsuits filed (Getty vs Stability AI, 2023).

> ⚖️ **THE QUESTION:** *Is training generative AI on copyrighted images without consent fair use?*

### 🟢 PRO It's Fair Use
1. **Transformation**: AI creates NEW content, not copies (legal fair use criterion)
2. **Photographers learn** from looking at thousands of photos — same principle
3. **Democratization**: people who can't afford artists now make beautiful images
4. **Innovation**: blocking AI would prevent next innovation (like color photography killed B&W painters)
5. **Open-source**: Stable Diffusion is free → benefits everyone, not just rich companies

### 🔴 CON Not Fair Use (Theft)
1. **No consent**: Getty's 12 million photos used without permission
2. **Direct competition**: AI image competing с original artist's livelihood — not transformation
3. **Style replication**: prompts like "in style of [living artist]" → unfair appropriation
4. **Scale matters**: human looks at 1000 images, AI memorizes 5 billion → quantitatively different
5. **Watermarks found**: Stable Diffusion images sometimes contain Getty watermarks → memorization, not learning

### 🔪 Killer Questions
**CON → PRO:** *"If I scan all your essays and create AI that writes 'in your style' — is that fair use? Why is artist work different?"*

**PRO → CON:** *"If we banned AI from learning from copyrighted work, we'd also ban art schools (they show students copyrighted paintings). Where's the line?"*

### 📚 Cases
- **Getty Images vs Stability AI** (2023 lawsuit)
- **Sarah Andersen vs DeviantArt** (artists' class action 2023)
- **EU AI Act** requires disclosure of training data
- **Japan**: AI training legal regardless of consent (controversial)
- **Adobe Firefly**: trained ONLY on licensed images (alternative model)

## 🎯 Debate #6: AI in Hiring — Efficiency or Discrimination?

### Scenario
> Корпорация получает 5000 резюме на одну vacancy. AI делает first filter за 5 секунд (vs 30 hours human HR). Но Amazon уже свернул свой recruiter в 2018 за discrimination.

> ⚖️ **THE QUESTION:** *Should companies be allowed to use AI for resume screening?*

### 🟢 PRO Allow
1. **Scale problem**: 5000 resumes × 30 min = 2500 hours. Impossible without AI
2. **Less human bias**: human recruiters discriminate by name, address, gender unconsciously (Harvard 2003 study)
3. **Better matches**: AI can find patterns humans miss (skills correlation, etc.)
4. **Audit trail**: AI decisions can be logged + reviewed; human decisions can't
5. **Already happens**: 75% of large companies use ATS (Applicant Tracking Systems)

### 🔴 CON Ban (or Strict regulation)
1. **Amazon 2018 lesson**: AI **discriminated against women** because trained on male-dominated past hires
2. **Bias multiplier**: ML doesn't reduce bias — it **amplifies** historical bias
3. **Black box**: rejected candidates **can't appeal**
4. **Proxy discrimination**: AI uses zip code, school → proxies for race/class
5. **EU AI Act** classifies hiring AI as "high-risk" → strict requirements coming

### 🔪 Killer Questions
**PRO → CON:** *"If we ban AI hiring, we go back to humans rejecting resumes for 'gut feeling'. Is that better?"*

**CON → PRO:** *"Amazon ran their AI for 4 years before noticing the bias. How can you guarantee yours won't have similar undetected bias?"*

### 📚 Cases
- **Amazon 2018** (scrapped AI recruiter)
- **HireVue** (AI video interview controversy)
- **Workday class action** (2024, alleged age discrimination)
- **EEOC guidance** (2023, on AI in hiring)
- **EU AI Act** high-risk classification

## 🎯 Debate #7: AI in Schools — Cheating Tool or Learning Aid?

### Scenario
> ChatGPT released Nov 2022 → schools panic. Some ban it. Others embrace. Students use it for homework, essays. Teachers can't reliably detect AI-written text.

> ⚖️ **THE QUESTION:** *Should schools allow students to use AI tools like ChatGPT?*

### 🟢 PRO Allow
1. **Real-world preparation**: students will use AI in their careers anyway
2. **Personalized tutor**: rich kids had tutors. AI = tutor for everyone (equity!)
3. **Higher-order thinking**: assignments become "critique this AI essay" → better learning
4. **Calculators precedent**: 1970s panic about calculators. Math education adapted, didn't collapse
5. **Productivity**: students focus on **understanding**, not on mechanical writing

### 🔴 CON Ban / Strict regulation
1. **Erodes critical thinking**: if AI does the thinking, brain atrophies (early studies show this)
2. **Cheating ≠ learning**: getting answer ≠ understanding
3. **Inequity**: rich kids get **better AI** (paid GPT-4, Claude Opus) → bigger gap
4. **Devalues education**: degrees mean less if AI did the work
5. **Foundation matters**: must learn basics **without** AI to use AI well later

### 🔪 Killer Questions
**PRO → CON:** *"You banned ChatGPT but allow Google. Where's the line? Both are tools that students could 'cheat' with."*

**CON → PRO:** *"Tom Hanks once said 'You learn nothing from your wins, you learn everything from your losses'. If AI does the hard parts, students never learn. How do you respond?"*

### 📚 Cases
- **NYC schools ChatGPT ban** (Jan 2023) → **reversed** May 2023
- **Australian universities**: hybrid approach (use AI but cite it)
- **Khan Academy Khanmigo**: school-approved AI tutor
- **Stanford pilots** with AI-assisted learning

## 🎯 Debate #8: Universal Basic Income — Should AI Profits Fund UBI?

### Scenario
> ML/AI eliminates jobs (truck drivers, paralegals, radiologists eventually). Yang 2020 presidential campaign popularized **UBI** ($1000/month). Should AI companies pay extra tax to fund UBI?

> ⚖️ **THE QUESTION:** *Should AI companies pay a special tax to fund Universal Basic Income for displaced workers?*

### 🟢 PRO UBI Tax
1. **Magnitude**: 47% of US jobs at risk (Oxford 2013 study); unprecedented disruption
2. **Concentration**: 5 companies (Google, Meta, Microsoft, OpenAI, Amazon) capturing most AI profits
3. **Social contract**: society **subsidized AI research** (universities, gov grants) — fair to recoup
4. **Stability**: mass unemployment → political extremism (history lesson: 1930s Germany)
5. **Pilot evidence**: Finland 2017-2018 UBI experiment showed positive mental health, productivity

### 🔴 CON No special tax
1. **Innovation killer**: special taxes → companies leave (Singapore, Dubai welcome them)
2. **What counts as "AI"?**: every software has some AI. Hard to define
3. **History of automation**: agricultural revolution displaced 80% farmers; new jobs emerged
4. **Disincentive to work**: UBI experiments mixed; some show decreased labor participation
5. **Better solutions**: education + retraining programs > UBI

### 🔪 Killer Questions
**PRO → CON:** *"Industrial Revolution displaced workers slowly over centuries. AI displaces them in years. How will society absorb this without intervention?"*

**CON → PRO:** *"If you tax AI heavily, all AI development moves to China. Are you prepared for AI hegemony by an authoritarian state?"*

### 📚 Cases
- **Andrew Yang 2020** Freedom Dividend ($1000/month)
- **Finland UBI experiment** 2017-2018
- **Sam Altman (OpenAI) UBI advocacy** (Worldcoin)
- **Oxford study** (Frey & Osborne, 2013)
- **Stockton SEED program** (California UBI pilot)


## 📊 PART 4 · Scoring Rubric (IB-Aligned)

> Судьи оценивают **каждую команду** по этой рубрике. Total = 30 баллов.

### Rubric (для печати судьям)

| Critère | 4 (excellent) | 3 (good) | 2 (acceptable) | 1 (weak) |
|---|---|---|---|---|
| **Argument quality** (×3) | Multiple strong, well-reasoned arguments with logic chains | 2-3 solid arguments | 1-2 basic arguments | Vague or contradictory |
| **Evidence & cases** (×3) | 4+ specific real cases with numbers/dates | 2-3 relevant cases | 1 case | No specific evidence |
| **Counter-argument handling** (×2) | Anticipates and refutes opposition skillfully | Addresses most opposing points | Addresses some | Ignores opposition |
| **Use of IB terminology** (×2) | Uses 5+ syllabus terms naturally (bias, accountability, GDPR, etc.) | 3-4 terms used | 1-2 terms | None |
| **Speaking & rhetoric** (×2) | Clear, persuasive, varied tone | Clear, organized | Mostly clear | Difficult to follow |
| **Teamwork** (×1) | Seamless coordination, equal participation | Good coordination | One person dominates | Disorganized |
| **Mitigation proposals** (×2) | Concrete, realistic mitigations proposed | Some mitigations | Vague mitigations | None |
| **TOTAL** | **/30** | | | |


## 🔗 PART 6 · Connection to IB IA

> Дебаты — это **золотая жила** для IA. Вот как использовать.

### Какой IA можно сделать из дебатов?

1. **Ethical Analysis IA**: возьмите **один** дебат → углубите → проведите interviews с stakeholders
2. **System Design with Ethics Review**: используйте **insights** из дебата для **dedicated ethics section** в вашем IA
3. **Mitigation Strategies Catalogue**: соберите **все** mitigation strategies из 8 дебатов → создайте **handbook** для разработчиков ML

### Структура IA с opening из debate

```
1. INTRODUCTION
   - Сформулируйте problem из debate
   - Контекст: "Topic was debated in our class on [date]"

2. STAKEHOLDER ANALYSIS
   - 3 группы интересов (из debate teams)

3. TECHNICAL SOLUTION
   - Ваш ML подход к проблеме

4. ETHICAL EVALUATION
   - PRO arguments (cite debate)
   - CON arguments (cite debate)
   - Mitigation matrix

5. CONCLUSION
   - Balanced recommendation
   - Future research needed
```

> 💎 **Pro tip:** мoderator/judges часто пишут лучшие IAs, потому что **слушали обе стороны** без bias. Если вы судья — используйте свои notes!
